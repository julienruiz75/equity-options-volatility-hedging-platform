## Motivation

Les options occupent une place centrale sur les marchés financiers, car leur valeur dépend simultanément de plusieurs facteurs de risque :

- le prix du sous-jacent,
- la volatilité,
- le temps restant jusqu’à l’échéance,
- les taux d’intérêt,
- et parfois les dividendes.

Contrairement aux produits linéaires, les options possèdent des **payoffs non linéaires**, ce qui rend leur profil de risque plus riche, plus complexe, et beaucoup plus intéressant à analyser.

L’idée centrale de ce notebook est la suivante :

> **Une option n’est pas seulement un prix : c’est une exposition dynamique à plusieurs facteurs de risque de marché.**

## Fondements mathématiques

Le cadre de Black-Scholes repose sur l’hypothèse que le prix du sous-jacent $S_t$ suit un mouvement brownien géométrique :

$$
dS_t = \mu S_t\,dt + \sigma S_t\,dW_t
$$

où :

- $\mu$ désigne la dérive,
- $\sigma$ la volatilité,
- $W_t$ un mouvement brownien.

En appliquant le lemme d’Itô à la valeur d’une option $V(S,t)$, puis en construisant un portefeuille auto-financé sans risque, on obtient l’équation aux dérivées partielles de Black-Scholes :

$$
\frac{\partial V}{\partial t}
+ \frac{1}{2}\sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}
+ rS \frac{\partial V}{\partial S}
- rV = 0
$$

Sous des conditions terminales appropriées, cette équation conduit aux formules fermées des options européennes.

Le prix d’un call européen s’écrit :

$$
C(S_0,K,T,r,\sigma) = S_0 \Phi(d_1) - K e^{-rT}\Phi(d_2)
$$

avec :

$$
d_1 = \frac{\ln\left(\frac{S_0}{K}\right) + \left(r + \frac{\sigma^2}{2}\right)T}{\sigma\sqrt{T}}
$$

$$
d_2 = d_1 - \sigma\sqrt{T}
$$

Le prix d’un put européen est donné par :

$$
P(S_0,K,T,r,\sigma) = K e^{-rT}\Phi(-d_2) - S_0 \Phi(-d_1)
$$

## Pourquoi les Greeks sont centraux

Une grande partie de ce notebook est consacrée aux **Greeks**, car ils traduisent les expositions au risque d’une option.

- Le **Delta** mesure la sensibilité du prix de l’option à une variation du sous-jacent.
- Le **Gamma** mesure la variation du Delta lorsque le sous-jacent évolue.
- Le **Vega** mesure la sensibilité à la volatilité.
- Le **Theta** mesure l’effet du temps sur la valeur de l’option.
- Le **Rho** mesure la sensibilité au taux d’intérêt.

En pratique, les Greeks sont essentiels pour :

- comprendre le profil de risque d’une position,
- mettre en place des couvertures,
- expliquer les variations quotidiennes de PnL,
- mesurer l’impact de chocs de marché.

Ainsi, ce notebook ne considère pas les Greeks comme un simple complément au pricing, mais comme le **langage central du risque optionnel**.
