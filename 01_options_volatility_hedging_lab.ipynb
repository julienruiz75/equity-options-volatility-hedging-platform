## Fondements mathématiques

Le cadre de Black-Scholes repose sur l’hypothèse que le prix du sous-jacent $S_t$ suit un mouvement brownien géométrique :

$$
dS_t = \mu S_t\,dt + \sigma S_t\,dW_t
$$

où :

- $\mu$ désigne la dérive,
- $\sigma$ la volatilité,
- $W_t$ un mouvement brownien.

En appliquant le lemme d’Itô à la valeur d’une option $V(S,t)$, puis en construisant un portefeuille auto-financé sans risque, on obtient l’équation aux dérivées partielles de Black-Scholes :

$$
\frac{\partial V}{\partial t}
+ \frac{1}{2}\sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}
+ rS \frac{\partial V}{\partial S}
- rV = 0
$$

Sous des conditions terminales appropriées, cette équation conduit aux formules fermées des options européennes.

Le prix d’un call européen s’écrit :

$$
C(S_0,K,T,r,\sigma) = S_0 \Phi(d_1) - K e^{-rT}\Phi(d_2)
$$

avec :

$$
d_1 = \frac{\ln\left(\frac{S_0}{K}\right) + \left(r + \frac{\sigma^2}{2}\right)T}{\sigma\sqrt{T}}
$$

$$
d_2 = d_1 - \sigma\sqrt{T}
$$

Le prix d’un put européen est donné par :

$$
P(S_0,K,T,r,\sigma) = K e^{-rT}\Phi(-d_2) - S_0 \Phi(-d_1)
$$

# 1. Fondements mathématiques

## 1.1 Pourquoi un cadre mathématique pour les options ?

Le prix d’une option dépend de plusieurs variables qui évoluent dans le temps :

- le prix du sous-jacent,
- la volatilité,
- le temps restant jusqu’à l’échéance,
- le taux d’intérêt,
- et parfois les dividendes.

Pour construire un modèle cohérent, il faut décrire mathématiquement l’évolution du sous-jacent, puis déterminer comment cette évolution affecte la valeur de l’option.

L’objectif de cette section est de poser les bases théoriques du modèle de Black-Scholes, qui servira ensuite pour :

- le pricing,
- le calcul des Greeks,
- la volatilité implicite,
- le hedging dynamique,
- et le stress testing.

---

## 1.2 Dynamique du sous-jacent

Dans le modèle de Black-Scholes, on suppose que le prix du sous-jacent $S_t$ suit un **mouvement brownien géométrique** :

$$
dS_t = \mu S_t\,dt + \sigma S_t\,dW_t
$$

où :

- $\mu$ est la dérive du sous-jacent,
- $\sigma$ est la volatilité,
- $W_t$ est un mouvement brownien standard.

Cette équation signifie que la variation instantanée du sous-jacent possède :

- une partie **déterministe** : $\mu S_t\,dt$,
- une partie **aléatoire** : $\sigma S_t\,dW_t$.

Le caractère multiplicatif du modèle permet de conserver un prix strictement positif, ce qui est cohérent pour un actif financier.

---

## 1.3 Valeur d’une option

On note $V(S,t)$ la valeur d’une option à la date $t$, lorsque le sous-jacent vaut $S$.

Comme $V$ dépend à la fois :

- du temps,
- et du niveau du sous-jacent,

on peut appliquer le **lemme d’Itô** à la fonction $V(S_t,t)$.

On obtient alors :

$$
dV = \frac{\partial V}{\partial t}dt
+ \frac{\partial V}{\partial S}dS
+ \frac{1}{2}\frac{\partial^2 V}{\partial S^2}(dS)^2
$$

En utilisant la dynamique de $S_t$ et la relation $(dW_t)^2 = dt$, cela devient :

$$
dV =
\left(
\frac{\partial V}{\partial t}
+ \mu S \frac{\partial V}{\partial S}
+ \frac{1}{2}\sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}
\right) dt
+ \sigma S \frac{\partial V}{\partial S} dW_t
$$

Cette expression est essentielle : elle montre que la valeur de l’option évolue elle aussi de façon stochastique.

---

## 1.4 Idée du portefeuille sans risque

L’intuition fondamentale de Black-Scholes consiste à construire un portefeuille combinant :

- une option,
- et une certaine quantité du sous-jacent,

de manière à éliminer le terme aléatoire en $dW_t$.

On considère le portefeuille :

$$
\Pi = V - \Delta S
$$

où $\Delta$ représente la quantité de sous-jacent détenue en couverture.

En choisissant :

$$
\Delta = \frac{\partial V}{\partial S}
$$

le terme stochastique disparaît.  
Le portefeuille devient alors localement sans risque.

Dans un marché sans opportunité d’arbitrage, un portefeuille sans risque doit rapporter le taux sans risque $r$.

Cette idée mène directement à l’équation de Black-Scholes.

---

## 1.5 Équation de Black-Scholes

En imposant la condition d’absence d’arbitrage, on obtient l’équation aux dérivées partielles suivante :

$$
\frac{\partial V}{\partial t}
+ \frac{1}{2}\sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}
+ rS \frac{\partial V}{\partial S}
- rV = 0
$$

Cette équation doit être satisfaite par toute option européenne dans le cadre du modèle de Black-Scholes.

Elle relie :

- la variation de la valeur dans le temps,
- la sensibilité au sous-jacent,
- la convexité,
- et le coût de financement au taux sans risque.

---

## 1.6 Conditions terminales

Pour résoudre l’équation de Black-Scholes, il faut spécifier le payoff à maturité.

Pour un **call européen** de strike $K$ et de maturité $T$ :

$$
V(S,T) = \max(S_T - K, 0)
$$

Pour un **put européen** :

$$
V(S,T) = \max(K - S_T, 0)
$$

À partir de ces conditions terminales, on obtient les formules fermées de Black-Scholes.

---

## 1.7 Formules de Black-Scholes

Le prix d’un call européen est donné par :

$$
C(S_0,K,T,r,\sigma) = S_0 \Phi(d_1) - K e^{-rT}\Phi(d_2)
$$

avec :

$$
d_1 = \frac{\ln\left(\frac{S_0}{K}\right) + \left(r + \frac{\sigma^2}{2}\right)T}{\sigma\sqrt{T}}
$$

$$
d_2 = d_1 - \sigma\sqrt{T}
$$

Le prix d’un put européen est :

$$
P(S_0,K,T,r,\sigma) = K e^{-rT}\Phi(-d_2) - S_0 \Phi(-d_1)
$$

où $\Phi$ désigne la fonction de répartition de la loi normale centrée réduite.

---

## 1.8 Intuition financière

Le modèle de Black-Scholes repose sur une idée très forte :

> il est possible de répliquer localement le risque d’une option à l’aide du sous-jacent.

Autrement dit, le prix d’une option n’est pas arbitraire : il découle d’un raisonnement de couverture et d’absence d’arbitrage.

C’est cette logique qui rend ensuite les **Greeks** si importants :

- le **Delta** vient naturellement de la couverture,
- le **Gamma** mesure la variation de cette couverture,
- le **Vega** capte l’effet de la volatilité,
- le **Theta** l’effet du temps,
- le **Rho** l’effet des taux.

Nous allons maintenant passer de ce cadre théorique à l’implémentation numérique du modèle.

# 2. Moteur de pricing Black-Scholes

Dans cette section, nous allons implémenter le moteur de pricing Black-Scholes pour les options européennes.

L’objectif est de construire des fonctions propres, réutilisables et robustes pour :

- calculer les quantités intermédiaires $d_1$ et $d_2$,
- valoriser un call européen,

```python
# ============================================================
# 2.0 Imports et configuration de base
# ============================================================

# -------------------------
# Calcul numérique et data
# -------------------------
import numpy as np
import pandas as pd

# -------------------------
# Statistiques / distributions
# -------------------------
from scipy.stats import norm
from scipy.optimize import brentq

# -------------------------
# Widgets et affichage notebook
# -------------------------
import ipywidgets as widgets
from IPython.display import display, HTML, Markdown

# -------------------------
# Visualisation Plotly
# -------------------------
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
from plotly.subplots import make_subplots

# -------------------------
# Réglages d'affichage pandas
# -------------------------
pd.set_option("display.max_columns", 50)
pd.set_option("display.width", 140)
pd.set_option("display.float_format", lambda x: f"{x:,.6f}")

# -------------------------
# Template graphique Plotly
# -------------------------
pio.templates.default = "plotly_dark"

print("Imports et configuration chargés avec succès.")
```

```python
# ============================================================
# 2.1 Fonctions de pricing Black-Scholes
# ============================================================

def _to_numpy(x):
    """
    Convertit une valeur scalaire ou un array-like en tableau numpy float.
    """
    return np.asarray(x, dtype=float)


def compute_d1(S, K, T, r, sigma):
    """
    Calcule d1 dans le modèle de Black-Scholes.
    """
    S = _to_numpy(S)
    K = _to_numpy(K)
    T = _to_numpy(T)
    r = _to_numpy(r)
    sigma = _to_numpy(sigma)

    if np.any(S <= 0):
        raise ValueError("Le spot S doit être strictement positif.")
    if np.any(K <= 0):
        raise ValueError("Le strike K doit être strictement positif.")
    if np.any(T <= 0):
        raise ValueError("La maturité T doit être strictement positive.")
    if np.any(sigma <= 0):
        raise ValueError("La volatilité sigma doit être strictement positive.")

    return (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))


def compute_d2(S, K, T, r, sigma):
    """
    Calcule d2 dans le modèle de Black-Scholes.
    """
    return compute_d1(S, K, T, r, sigma) - sigma * np.sqrt(T)


def bs_call_price(S, K, T, r, sigma):
    """
    Prix d'un call européen dans le modèle de Black-Scholes.
    """
    d1 = compute_d1(S, K, T, r, sigma)
    d2 = compute_d2(S, K, T, r, sigma)
    return S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)


def bs_put_price(S, K, T, r, sigma):
    """
    Prix d'un put européen dans le modèle de Black-Scholes.
    """
    d1 = compute_d1(S, K, T, r, sigma)
    d2 = compute_d2(S, K, T, r, sigma)
    return K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
```

```python
# ============================================================
# 2.2 Fonctions des Greeks Black-Scholes
# ============================================================

def call_delta(S, K, T, r, sigma):
    """
    Delta d'un call européen.
    """
    d1 = compute_d1(S, K, T, r, sigma)
    return norm.cdf(d1)


def put_delta(S, K, T, r, sigma):
    """
    Delta d'un put européen.
    """
    d1 = compute_d1(S, K, T, r, sigma)
    return norm.cdf(d1) - 1


def gamma_bs(S, K, T, r, sigma):
    """
    Gamma d'un call ou d'un put européen.
    Identique pour call et put dans Black-Scholes.
    """
    d1 = compute_d1(S, K, T, r, sigma)
    S = _to_numpy(S)
    sigma = _to_numpy(sigma)
    T = _to_numpy(T)

    return norm.pdf(d1) / (S * sigma * np.sqrt(T))


def vega_bs(S, K, T, r, sigma):
    """
    Vega d'un call ou d'un put européen.
    """
    d1 = compute_d1(S, K, T, r, sigma)
    S = _to_numpy(S)
    T = _to_numpy(T)

    return S * norm.pdf(d1) * np.sqrt(T)


def call_theta(S, K, T, r, sigma):
    """
    Theta d'un call européen.
    """
    d1 = compute_d1(S, K, T, r, sigma)
    d2 = compute_d2(S, K, T, r, sigma)

    S = _to_numpy(S)
    K = _to_numpy(K)
    T = _to_numpy(T)
    sigma = _to_numpy(sigma)

    term1 = -S * norm.pdf(d1) * sigma / (2 * np.sqrt(T))
    term2 = -r * K * np.exp(-r * T) * norm.cdf(d2)

    return term1 + term2


def put_theta(S, K, T, r, sigma):
    """
    Theta d'un put européen.
    """
    d1 = compute_d1(S, K, T, r, sigma)
    d2 = compute_d2(S, K, T, r, sigma)

    S = _to_numpy(S)
    K = _to_numpy(K)
    T = _to_numpy(T)
    sigma = _to_numpy(sigma)

    term1 = -S * norm.pdf(d1) * sigma / (2 * np.sqrt(T))
    term2 = r * K * np.exp(-r * T) * norm.cdf(-d2)

    return term1 + term2


def call_rho(S, K, T, r, sigma):
    """
    Rho d'un call européen.
    """
    d2 = compute_d2(S, K, T, r, sigma)
    K = _to_numpy(K)
    T = _to_numpy(T)

    return K * T * np.exp(-r * T) * norm.cdf(d2)


def put_rho(S, K, T, r, sigma):
    """
    Rho d'un put européen.
    """
    d2 = compute_d2(S, K, T, r, sigma)
    K = _to_numpy(K)
    T = _to_numpy(T)

    return -K * T * np.exp(-r * T) * norm.cdf(-d2)
```

```python
# ============================================================
# 2.3 Panneau interactif de pricing + Greeks Black-Scholes
# ============================================================

import ipywidgets as widgets
from IPython.display import display, HTML

# -------------------------
# Widgets de saisie
# -------------------------
option_widget = widgets.Dropdown(
    options=[("Call européen", "call"), ("Put européen", "put")],
    value="call",
    description="Option",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="260px")
)

S_widget = widgets.FloatText(
    value=100.0,
    description="Spot S",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="260px")
)

K_widget = widgets.FloatText(
    value=100.0,
    description="Strike K",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="260px")
)

T_widget = widgets.FloatText(
    value=1.0,
    description="Maturité T",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="260px")
)

r_widget = widgets.FloatText(
    value=0.03,
    description="Taux r",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="260px")
)

sigma_widget = widgets.FloatText(
    value=0.20,
    description="Volatilité σ",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="260px")
)

output = widgets.Output()

# -------------------------
# Fonction utilitaire
# -------------------------
def classify_moneyness(option_type, S, K, tol=0.01):
    rel_diff = abs(S - K) / K
    if rel_diff <= tol:
        return "ATM"
    if option_type == "call":
        return "ITM" if S > K else "OTM"
    return "ITM" if S < K else "OTM"

# -------------------------
# Fonction principale
# -------------------------
def update_pricer(*args):
    with output:
        output.clear_output(wait=True)

        option_type = option_widget.value
        S = S_widget.value
        K = K_widget.value
        T = T_widget.value
        r = r_widget.value
        sigma = sigma_widget.value

        try:
            d1 = float(compute_d1(S, K, T, r, sigma))
            d2 = float(compute_d2(S, K, T, r, sigma))

            if option_type == "call":
                price = float(bs_call_price(S, K, T, r, sigma))
                delta = float(call_delta(S, K, T, r, sigma))
                theta = float(call_theta(S, K, T, r, sigma))
                rho = float(call_rho(S, K, T, r, sigma))
                option_label = "Call européen"
                intrinsic_value = max(S - K, 0.0)
            else:
                price = float(bs_put_price(S, K, T, r, sigma))
                delta = float(put_delta(S, K, T, r, sigma))
                theta = float(put_theta(S, K, T, r, sigma))
                rho = float(put_rho(S, K, T, r, sigma))
                option_label = "Put européen"
                intrinsic_value = max(K - S, 0.0)

            gamma_value = float(gamma_bs(S, K, T, r, sigma))
            vega_value = float(vega_bs(S, K, T, r, sigma))

            time_value = price - intrinsic_value
            moneyness = classify_moneyness(option_type, S, K)

            html = f"""
            <div style="font-family: Arial, sans-serif; max-width: 1200px;">

                <div style="margin: 18px 0 10px 0;">
                    <h3 style="margin: 0; color: #1f1f1f;">Résultats du pricing et des Greeks</h3>
                    <p style="margin: 6px 0 0 0; color: #555;">
                        Prix théorique Black-Scholes et sensibilités principales de l'option.
                    </p>
                </div>

                <div style="display: grid; grid-template-columns: repeat(4, 1fr); gap: 14px; margin-bottom: 20px;">
                    <div style="background: #f5f7fb; border: 1px solid #d9e2f1; border-radius: 12px; padding: 14px;">
                        <div style="font-size: 13px; color: #6b7280;">Type d'option</div>
                        <div style="font-size: 20px; font-weight: 700; margin-top: 6px;">{option_label}</div>
                    </div>

                    <div style="background: #f5f7fb; border: 1px solid #d9e2f1; border-radius: 12px; padding: 14px;">
                        <div style="font-size: 13px; color: #6b7280;">Prix Black-Scholes</div>
                        <div style="font-size: 20px; font-weight: 700; margin-top: 6px;">{price:.6f}</div>
                    </div>

                    <div style="background: #f5f7fb; border: 1px solid #d9e2f1; border-radius: 12px; padding: 14px;">
                        <div style="font-size: 13px; color: #6b7280;">Moneyness</div>
                        <div style="font-size: 20px; font-weight: 700; margin-top: 6px;">{moneyness}</div>
                    </div>

                    <div style="background: #f5f7fb; border: 1px solid #d9e2f1; border-radius: 12px; padding: 14px;">
                        <div style="font-size: 13px; color:px; color: #6b7280;">Valeur temps</div>
                        <div style="font-size: 20px; font-weight: 700; margin-top: 6px;">{time_value:.6f}</div>
                    </div>
                </div>

                <div style="display: grid; grid-template-columns: repeat(5, 1fr); gap: 14px; margin-bottom: 22px;">
                    <div style="background: #fff8ef; border: 1px solid #f0d9b5; border-radius: 12px; padding: 14px;">
                        <div style="font-size: 13px; color: #7a5c2e;">Delta</div>
                        <div style="font-size: 22px; font-weight: 700; margin-top: 6px;">{delta:.6f}</div>
                    </div>

                    <div style="background: #fff8ef; border: 1px solid #f0d9b5; border-radius: 12px; padding: 14px;">
                        <div style="font-size: 13px; color: #7a5c2e;">Gamma</div>
                        <div style="font-size: 22px; font-weight: 700; margin-top: 6px;">{gamma_value:.6f}</div>
                    </div>

                    <div style="background: #fff8ef; border: 1px solid #f0d9b5; border-radius: 12px; padding: 14px;">
                        <div style="font-size: 13px; color: #7a5c2e;">Vega</div>
                        <div style="font-size: 22px; font-weight: 700; margin-top: 6px;">{vega_value:.6f}</div>
                    </div>

                    <div style="background: #fff8ef; border: 1px solid #f0d9b5; border-radius: 12px; padding: 14px;">
                        <div style="font-size: 13px; color: #7a5c2e;">Theta</div>
                        <div style="font-size: 22px; font-weight: 700; margin-top: 6px;">{theta:.6f}</div>
                    </div>

                    <div style="background: #fff8ef; border: 1px solid #f0d9b5; border-radius: 12px; padding: 14px;">
                        <div style="font-size: 13px; color: #7a5c2e;">Rho</div>
                        <div style="font-size: 22px; font-weight: 700; margin-top: 6px;">{rho:.6f}</div>
                    </div>
                </div>

                <table style="border-collapse: collapse; width: 100%; background: white; border-radius: 12px; overflow: hidden;">
                    <thead>
                        <tr style="background: #eef2f7;">
                            <th style="text-align: left; padding: 12px; border-bottom: 1px solid #d9e2f1;">Paramètre</th>
                            <th style="text-align: right; padding: 12px; border-bottom: 1px solid #d9e2f1;">Valeur</th>
                        </tr>
                    </thead>
                    <tbody>
                        <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Spot</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{S:.6f}</td></tr>
                        <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Strike</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{K:.6f}</td></tr>
                        <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Maturité (années)</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{T:.6f}</td></tr>
                        <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Taux sans risque</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{r:.6%}</td></tr>
                        <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Volatilité</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{sigma:.6%}</td></tr>
                        <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">d1</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{d1:.6f}</td></tr>
                        <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">d2</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{d2:.6f}</td></tr>
                        <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Valeur intrinsèque</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{intrinsic_value:.6f}</td></tr>
                        <tr><td style="padding: 10px 12px;">Valeur temps</td><td style="padding: 10px 12px; text-align: right;">{time_value:.6f}</td></tr>
                    </tbody>
                </table>
            </div>
            """

            display(HTML(html))

        except Exception as e:
            display(HTML(f"<b style='color:#d62828;'>Erreur :</b> {e}"))

# -------------------------
# Rafraîchissement auto
# -------------------------
for w in [option_widget, S_widget, K_widget, T_widget, r_widget, sigma_widget]:
    w.observe(update_pricer, names="value")

# -------------------------
# Mise en page
# -------------------------
input_grid = widgets.GridBox(
    children=[
        option_widget,
        S_widget,
        K_widget,
        T_widget,
        r_widget,
        sigma_widget
    ],
    layout=widgets.Layout(
        grid_template_columns="repeat(3, 280px)",
        grid_gap="12px 18px",
        margin="10px 0 20px 0"
    )
)

ui = widgets.VBox([
    widgets.HTML("<h3 style='margin:0;'>Panneau interactif de pricing Black-Scholes</h3>"),
    widgets.HTML("<p style='margin:6px 0 8px 0; color:#555;'>Renseignez les paramètres de l'option pour obtenir instantanément son prix théorique et ses sensibilités.</p>"),
    input_grid
])

display(ui, output)

# Premier affichage
update_pricer()

# ============================================================
# 2.4 Graphique dynamique des sensibilités piloté par le panneau
# version plus propre, sans texte superposé sur le graphe
# ============================================================

greek_widget = widgets.Dropdown(
    options=[
        ("Delta", "delta"),
        ("Gamma", "gamma"),
        ("Vega", "vega"),
        ("Theta", "theta"),
        ("Rho", "rho"),
        ("Prix", "price")
    ],
    value="delta",
    description="Courbe",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="260px")
)

graph_output = widgets.Output()

def update_dynamic_chart(*args):
    with graph_output:
        graph_output.clear_output(wait=True)

        option_type = option_widget.value
        S = S_widget.value
        K = K_widget.value
        T = T_widget.value
        r = r_widget.value
        sigma = sigma_widget.value
        selected_curve = greek_widget.value

        try:
            # Grille de spot centrée autour du strike
            spot_min = max(1e-6, 0.5 * K)
            spot_max = 1.5 * K
            spot_range = np.linspace(spot_min, spot_max, 400)

            # Choix de la courbe à afficher
            if selected_curve == "price":
                if option_type == "call":
                    y_values = bs_call_price(spot_range, K, T, r, sigma)
                    current_value = float(bs_call_price(S, K, T, r, sigma))
                    curve_name = "Prix du call"
                else:
                    y_values = bs_put_price(spot_range, K, T, r, sigma)
                    current_value = float(bs_put_price(S, K, T, r, sigma))
                    curve_name = "Prix du put"

            elif selected_curve == "delta":
                if option_type == "call":
                    y_values = call_delta(spot_range, K, T, r, sigma)
                    current_value = float(call_delta(S, K, T, r, sigma))
                    curve_name = "Delta du call"
                else:
                    y_values = put_delta(spot_range, K, T, r, sigma)
                    current_value = float(put_delta(S, K, T, r, sigma))
                    curve_name = "Delta du put"

            elif selected_curve == "gamma":
                y_values = gamma_bs(spot_range, K, T, r, sigma)
                current_value = float(gamma_bs(S, K, T, r, sigma))
                curve_name = "Gamma"

            elif selected_curve == "vega":
                y_values = vega_bs(spot_range, K, T, r, sigma)
                current_value = float(vega_bs(S, K, T, r, sigma))
                curve_name = "Vega"

            elif selected_curve == "theta":
                if option_type == "call":
                    y_values = call_theta(spot_range, K, T, r, sigma) / 365.0
                    current_value = float(call_theta(S, K, T, r, sigma) / 365.0)
                    curve_name = "Theta du call (par jour)"
                else:
                    y_values = put_theta(spot_range, K, T, r, sigma) / 365.0
                    current_value = float(put_theta(S, K, T, r, sigma) / 365.0)
                    curve_name = "Theta du put (par jour)"

            elif selected_curve == "rho":
                if option_type == "call":
                    y_values = call_rho(spot_range, K, T, r, sigma) * 0.01
                    current_value = float(call_rho(S, K, T, r, sigma) * 0.01)
                    curve_name = "Rho du call (pour +1%)"
                else:
                    y_values = put_rho(spot_range, K, T, r, sigma) * 0.01
                    current_value = float(put_rho(S, K, T, r, sigma) * 0.01)
                    curve_name = "Rho du put (pour +1%)"

            else:
                raise ValueError("Courbe non reconnue.")

            # Bandeau d'information au-dessus du graphe
            info_html = f"""
            <div style="display:flex; gap:12px; flex-wrap:wrap; margin:8px 0 14px 0; font-family:Arial, sans-serif;">
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                    <div style="font-size:12px; color:#6b7280;">Courbe affichée</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{curve_name}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:140px;">
                    <div style="font-size:12px; color:#6b7280;">Spot actuel</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{S:.2f}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:140px;">
                    <div style="font-size:12px; color:#6b7280;">Strike</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{K:.2f}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                    <div style="font-size:12px; color:#6b7280;">Valeur actuelle</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{current_value:.6f}</div>
                </div>
            </div>
            """
            display(HTML(info_html))

            fig = go.Figure()

            fig.add_trace(go.Scatter(
                x=spot_range,
                y=y_values,
                mode="lines",
                name=curve_name,
                line=dict(width=3)
            ))

            # Ligne strike sans annotation text
            fig.add_shape(
                type="line",
                x0=K, x1=K,
                y0=float(np.min(y_values)), y1=float(np.max(y_values)),
                line=dict(color="white", width=1.2, dash="dash")
            )

            # Ligne spot actuel sans annotation text
            fig.add_shape(
                type="line",
                x0=S, x1=S,
                y0=float(np.min(y_values)), y1=float(np.max(y_values)),
                line=dict(color="#00d4ff", width=1.2, dash="dot")
            )

            # Point courant
            current_y = np.interp(S, spot_range, y_values)
            fig.add_trace(go.Scatter(
                x=[S],
                y=[current_y],
                mode="markers",
                name="Point courant",
                marker=dict(size=10, color="white", line=dict(color="black", width=1)),
                hovertemplate=(
                    f"Spot actuel = {S:.2f}<br>"
                    f"{curve_name} = {current_y:.6f}<extra></extra>"
                )
            ))

            fig.update_layout(
                title=f"{curve_name} en fonction du spot",
                xaxis_title="Spot",
                yaxis_title=curve_name,
                template="plotly_dark",
                hovermode="x unified",
                height=550,
                margin=dict(l=70, r=40, t=70, b=70),
                legend=dict(
                    orientation="h",
                    yanchor="bottom",
                    y=1.02,
                    xanchor="right",
                    x=1
                )
            )

            fig.show()

        except Exception as e:
            display(HTML(f"<b style='color:#d62828;'>Erreur graphique :</b> {e}"))

for w in [option_widget, S_widget, K_widget, T_widget, r_widget, sigma_widget, greek_widget]:
    w.observe(update_dynamic_chart, names="value")

chart_ui = widgets.VBox([
    widgets.HTML("<h3 style='margin:16px 0 6px 0;'>Analyse graphique dynamique</h3>"),
    widgets.HTML("<p style='margin:0 0 10px 0; color:#555;'>Choisissez la courbe à afficher. Le graphique se met automatiquement à jour avec les paramètres du panneau.</p>"),
    greek_widget
])

display(chart_ui, graph_output)
update_dynamic_chart()

# ============================================================
# 2.5 Heatmap 2D pilotée par le panneau : version corrigée
# avec bandeau d'information + valeur actuelle de la métrique
# ============================================================

surface_widget = widgets.Dropdown(
    options=[
        ("Prix", "price"),
        ("Delta", "delta"),
        ("Gamma", "gamma"),
        ("Vega", "vega"),
        ("Theta", "theta"),
        ("Rho", "rho")
    ],
    value="gamma",
    description="Heatmap",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="260px")
)

heatmap_output = widgets.Output()

def update_heatmap(*args):
    with heatmap_output:
        heatmap_output.clear_output(wait=True)

        option_type = option_widget.value
        S = S_widget.value
        K = K_widget.value
        T = T_widget.value
        r = r_widget.value
        sigma = sigma_widget.value
        selected_surface = surface_widget.value

        try:
            # -------------------------
            # Grille Spot x Volatilité
            # -------------------------
            spot_min = max(1e-6, 0.5 * K)
            spot_max = 1.5 * K
            vol_min = 0.05
            vol_max = max(0.80, sigma * 2)

            spot_range = np.linspace(spot_min, spot_max, 140)
            vol_range = np.linspace(vol_min, vol_max, 140)

            S_mesh, sigma_mesh = np.meshgrid(spot_range, vol_range)

            # -------------------------
            # Choix de la métrique
            # -------------------------
            if selected_surface == "price":
                if option_type == "call":
                    Z = bs_call_price(S_mesh, K, T, r, sigma_mesh)
                    current_value = float(bs_call_price(S, K, T, r, sigma))
                    title_metric = "Prix du call"
                else:
                    Z = bs_put_price(S_mesh, K, T, r, sigma_mesh)
                    current_value = float(bs_put_price(S, K, T, r, sigma))
                    title_metric = "Prix du put"
                colorscale = "Cividis"
                zmid = None

            elif selected_surface == "delta":
                if option_type == "call":
                    Z = call_delta(S_mesh, K, T, r, sigma_mesh)
                    current_value = float(call_delta(S, K, T, r, sigma))
                    title_metric = "Delta du call"
                else:
                    Z = put_delta(S_mesh, K, T, r, sigma_mesh)
                    current_value = float(put_delta(S, K, T, r, sigma))
                    title_metric = "Delta du put"
                colorscale = "RdBu"
                zmid = 0

            elif selected_surface == "gamma":
                Z = gamma_bs(S_mesh, K, T, r, sigma_mesh)
                current_value = float(gamma_bs(S, K, T, r, sigma))
                title_metric = "Gamma"
                colorscale = "Viridis"
                zmid = None

            elif selected_surface == "vega":
                Z = vega_bs(S_mesh, K, T, r, sigma_mesh)
                current_value = float(vega_bs(S, K, T, r, sigma))
                title_metric = "Vega"
                colorscale = "Plasma"
                zmid = None

            elif selected_surface == "theta":
                if option_type == "call":
                    Z = call_theta(S_mesh, K, T, r, sigma_mesh) / 365.0
                    current_value = float(call_theta(S, K, T, r, sigma) / 365.0)
                    title_metric = "Theta du call (par jour)"
                else:
                    Z = put_theta(S_mesh, K, T, r, sigma_mesh) / 365.0
                    current_value = float(put_theta(S, K, T, r, sigma) / 365.0)
                    title_metric = "Theta du put (par jour)"
                colorscale = "RdBu"
                zmid = 0

            elif selected_surface == "rho":
                if option_type == "call":
                    Z = call_rho(S_mesh, K, T, r, sigma_mesh) * 0.01
                    current_value = float(call_rho(S, K, T, r, sigma) * 0.01)
                    title_metric = "Rho du call (pour +1%)"
                else:
                    Z = put_rho(S_mesh, K, T, r, sigma_mesh) * 0.01
                    current_value = float(put_rho(S, K, T, r, sigma) * 0.01)
                    title_metric = "Rho du put (pour +1%)"
                colorscale = "RdBu"
                zmid = 0

            else:
                raise ValueError("Type de heatmap non reconnu.")

            # -------------------------
            # Bandeau d'information
            # -------------------------
            info_html = f"""
            <div style="display:flex; gap:12px; flex-wrap:wrap; margin:8px 0 14px 0; font-family:Arial, sans-serif;">
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                    <div style="font-size:12px; color:#6b7280;">Métrique affichée</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{title_metric}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                    <div style="font-size:12px; color:#6b7280;">Spot actuel</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{S:.2f}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                    <div style="font-size:12px; color:#6b7280;">Vol actuelle</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{sigma:.2%}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                    <div style="font-size:12px; color:#6b7280;">Strike</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{K:.2f}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                    <div style="font-size:12px; color:#6b7280;">Valeur actuelle</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{current_value:.6f}</div>
                </div>
            </div>
            """
            display(HTML(info_html))

            # -------------------------
            # Construction de la figure
            # -------------------------
            fig = go.Figure()

            fig.add_trace(go.Heatmap(
                x=spot_range,
                y=vol_range,
                z=Z,
                colorscale=colorscale,
                zmid=zmid,
                colorbar=dict(
                    title=title_metric,
                    thickness=18,
                    len=0.78,
                    y=0.5
                ),
                hovertemplate=(
                    "Spot : %{x:.2f}<br>"
                    "Volatilité : %{y:.2%}<br>"
                    f"{title_metric} : " + "%{z:.6f}<extra></extra>"
                )
            ))

            # Strike
            fig.add_shape(
                type="line",
                x0=K, x1=K,
                y0=vol_min, y1=vol_max,
                line=dict(color="white", width=1.2, dash="dash")
            )

            # Spot actuel
            fig.add_shape(
                type="line",
                x0=S, x1=S,
                y0=vol_min, y1=vol_max,
                line=dict(color="#00d4ff", width=1.2, dash="dot")
            )

            # Vol actuelle
            fig.add_shape(
                type="line",
                x0=spot_min, x1=spot_max,
                y0=sigma, y1=sigma,
                line=dict(color="#ffcc00", width=1.2, dash="dot")
            )

            # Point courant
            fig.add_trace(go.Scatter(
                x=[S],
                y=[sigma],
                mode="markers",
                name="Point courant",
                marker=dict(
                    size=10,
                    color="white",
                    line=dict(color="black", width=1)
                ),
                hovertemplate=(
                    "Point courant<br>"
                    f"Spot = {S:.2f}<br>"
                    f"Vol = {sigma:.2%}<br>"
                    f"{title_metric} = {current_value:.6f}<extra></extra>"
                )
            ))

            fig.update_layout(
                title=dict(
                    text=f"Heatmap {title_metric} — Spot × Volatilité",
                    x=0.5,
                    xanchor="center"
                ),
                template="plotly_dark",
                height=650,
                margin=dict(l=70, r=40, t=70, b=70),
                legend=dict(
                    orientation="h",
                    yanchor="bottom",
                    y=1.02,
                    xanchor="right",
                    x=1
                )
            )

            fig.update_xaxes(
                title="Spot",
                showgrid=False,
                zeroline=False
            )

            fig.update_yaxes(
                title="Volatilité",
                tickformat=".0%",
                showgrid=False,
                zeroline=False
            )

            fig.show()

        except Exception as e:
            display(HTML(f"<b style='color:#d62828;'>Erreur heatmap :</b> {e}"))

for w in [option_widget, S_widget, K_widget, T_widget, r_widget, sigma_widget, surface_widget]:
    w.observe(update_heatmap, names="value")

heatmap_ui = widgets.VBox([
    widgets.HTML("<h3 style='margin:16px 0 6px 0;'>Heatmap 2D des sensibilités</h3>"),
    widgets.HTML("<p style='margin:0 0 10px 0; color:#555;'>Visualisation de la métrique choisie sur une grille Spot × Volatilité, pilotée par le panneau principal.</p>"),
    surface_widget
])

display(heatmap_ui, heatmap_output)
update_heatmap()
```

# 3. Données de marché et volatilité implicite

Jusqu’à présent, le notebook a développé un cadre théorique de pricing et d’analyse des sensibilités dans le modèle de Black-Scholes.

La prochaine étape consiste à confronter ce cadre aux **prix réellement observés sur le marché**.

L’objectif est triple :

- récupérer une chaîne d’options réelle sur un sous-jacent coté,
- comparer les prix de marché aux prix théoriques Black-Scholes,
- calculer la **volatilité implicite**, c’est-à-dire la volatilité qui permet au modèle de retrouver le prix observé.

Cette section constitue un tournant important dans le projet, car elle fait passer l’analyse :

- d’un cadre purement théorique,
- à une lecture concrète du marché des options.

Nous commencerons par charger les données d’un sous-jacent, lister les maturités disponibles, puis sélectionner une échéance pour analyser sa chaîne d’options.

```python
# ============================================================
# 3.1 Panneau marché unifié complet
# sous-jacent + échéance + chaîne + contrat + IV + smile
# + structure par terme + surface 3D interpolée
# + smiles multi-maturités + mispricing
# ============================================================

import numpy as np
import pandas as pd
import ipywidgets as widgets
from IPython.display import display, HTML
import plotly.graph_objects as go
import yfinance as yf
from scipy.optimize import brentq

market_ticker_widget = widgets.Text(
    value="AAPL",
    description="Ticker",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="220px")
)

load_market_button = widgets.Button(
    description="Charger",
    button_style="primary",
    layout=widgets.Layout(width="140px")
)

expiry_widget = widgets.Dropdown(
    options=[],
    description="Échéance",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="220px")
)

option_side_widget = widgets.Dropdown(
    options=[("Calls", "calls"), ("Puts", "puts")],
    value="calls",
    description="Chaîne",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="180px")
)

rows_widget = widgets.IntSlider(
    value=12,
    min=5,
    max=30,
    step=1,
    description="Lignes",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="280px")
)

contract_strike_widget = widgets.Dropdown(
    options=[],
    description="Strike",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="220px")
)

price_source_widget = widgets.Dropdown(
    options=[("Mid price", "midPrice"), ("Last price", "lastPrice")],
    value="midPrice",
    description="Prix mkt",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="220px")
)

term_expiries_widget = widgets.IntSlider(
    value=10,
    min=3,
    max=20,
    step=1,
    description="Maturités",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="280px")
)

surface_expiries_widget = widgets.IntSlider(
    value=8,
    min=3,
    max=15,
    step=1,
    description="Éch. 3D",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="260px")
)

compare_expiries_widget = widgets.IntSlider(
    value=4,
    min=2,
    max=8,
    step=1,
    description="Smiles",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="240px")
)

surface_metric_widget = widgets.Dropdown(
    options=[("IV Yahoo", "iv_yahoo"), ("IV maison", "iv_home")],
    value="iv_yahoo",
    description="Surface",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="220px")
)

strike_window_widget = widgets.FloatSlider(
    value=0.25,
    min=0.10,
    max=0.50,
    step=0.05,
    description="Fenêtre K",
    readout_format=".0%",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="260px")
)

min_volume_widget = widgets.IntSlider(
    value=0,
    min=0,
    max=5000,
    step=50,
    description="Vol min",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="240px")
)

min_oi_widget = widgets.IntSlider(
    value=0,
    min=0,
    max=10000,
    step=50,
    description="OI min",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="240px")
)

mispricing_top_widget = widgets.IntSlider(
    value=12,
    min=5,
    max=25,
    step=1,
    description="Top écarts",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="260px")
)

market_output = widgets.Output()

market_state = {
    "symbol": None,
    "ticker": None,
    "spot": None,
    "expirations": [],
    "calls": None,
    "puts": None,
    "expiry": None,
    "chain_cache": {},
    "is_updating": False
}

# ------------------------------------------------------------
# Fonctions utilitaires
# ------------------------------------------------------------
def fmt_num(x, pct=False):
    if pd.isna(x):
        return "NA"
    return f"{x:.6%}" if pct else f"{x:.6f}"

def fmt_int(x):
    if pd.isna(x):
        return "NA"
    return f"{int(x)}"

def get_model_params():
    try:
        r_value = float(r_widget.value)
    except Exception:
        r_value = 0.03

    try:
        sigma_value = float(sigma_widget.value)
    except Exception:
        sigma_value = 0.20

    return r_value, sigma_value

def render_dataframe_html(df):
    if df is None or df.empty:
        return "<p style='color:#555; font-family:Arial, sans-serif;'>Aucune donnée à afficher.</p>"

    table_html = df.to_html(index=False, border=0)

    styled_table_html = f"""
    <div style="font-family: Arial, sans-serif; margin-top: 8px; overflow-x:auto;">
        <style>
            .custom-market-table {{
                border-collapse: collapse;
                width: 100%;
                background: white;
                border-radius: 12px;
                overflow: hidden;
            }}
            .custom-market-table th {{
                background: #eef2f7;
                padding: 10px 12px;
                text-align: center;
                border-bottom: 1px solid #d9e2f1;
                white-space: nowrap;
            }}
            .custom-market-table td {{
                padding: 10px 12px;
                text-align: center;
                border-bottom: 1px solid #edf1f7;
                white-space: nowrap;
            }}
            .custom-market-table tr:last-child td {{
                border-bottom: none;
            }}
        </style>
        {table_html.replace('<table border="0" class="dataframe">', '<table class="custom-market-table">')}
    </div>
    """
    return styled_table_html

def get_current_chain_df():
    if market_state["calls"] is None or market_state["puts"] is None:
        return None
    return market_state["calls"].copy() if option_side_widget.value == "calls" else market_state["puts"].copy()

def compute_mid_price(df):
    if "bid" in df.columns and "ask" in df.columns:
        df["midPrice"] = (df["bid"].fillna(0) + df["ask"].fillna(0)) / 2
    return df

def preprocess_chain_side(df, spot):
    useful_cols = [
        "contractSymbol",
        "strike",
        "lastPrice",
        "bid",
        "ask",
        "impliedVolatility",
        "volume",
        "openInterest",
        "inTheMoney"
    ]
    useful_cols = [c for c in useful_cols if c in df.columns]
    df = df[useful_cols].copy()
    df = compute_mid_price(df)

    if "strike" in df.columns:
        df["distance_to_spot"] = (df["strike"] - spot).abs()
        df = df.sort_values("distance_to_spot").drop(columns=["distance_to_spot"])

    return df

def refresh_strike_choices(processed_df=None):
    if processed_df is None:
        df = get_current_chain_df()
        if df is None:
            contract_strike_widget.options = []
            return
        spot = market_state["spot"]
        df = preprocess_chain_side(df, spot)
    else:
        df = processed_df

    if df is None or "strike" not in df.columns or df.empty:
        contract_strike_widget.options = []
        return

    strikes = sorted(df["strike"].dropna().unique().tolist())
    new_options = [(f"{s:.2f}", float(s)) for s in strikes]
    previous_value = contract_strike_widget.value

    current_options = list(contract_strike_widget.options)
    if current_options != new_options:
        contract_strike_widget.options = new_options

    if previous_value in strikes:
        contract_strike_widget.value = float(previous_value)
    else:
        spot = market_state["spot"]
        if spot is not None and len(strikes) > 0:
            nearest_strike = min(strikes, key=lambda x: abs(x - spot))
            contract_strike_widget.value = float(nearest_strike)

def get_market_price_from_row(row, preferred_source):
    market_price = row.get(preferred_source, np.nan)
    actual_price_source = preferred_source

    if pd.isna(market_price) or market_price == 0:
        market_price = row.get("lastPrice", np.nan)
        actual_price_source = "lastPrice"

    return market_price, actual_price_source

def get_chain_for_expiry(ticker, expiry):
    if expiry in market_state["chain_cache"]:
        return market_state["chain_cache"][expiry]

    chain = ticker.option_chain(expiry)
    calls_df = chain.calls.copy()
    puts_df = chain.puts.copy()

    market_state["chain_cache"][expiry] = (calls_df, puts_df)
    return calls_df, puts_df

def classify_moneyness(option_label, spot, strike, tol=0.01):
    rel_diff = abs(spot - strike) / strike
    if rel_diff <= tol:
        return "ATM"
    if option_label == "Call":
        return "ITM" if spot > strike else "OTM"
    return "ITM" if spot < strike else "OTM"

def filter_liquid_contracts(df, strike_min, strike_max, min_volume, min_oi):
    filtered = df.copy()
    filtered = filtered[
        (filtered["strike"] >= strike_min) &
        (filtered["strike"] <= strike_max)
    ].copy()

    if "volume" in filtered.columns:
        filtered = filtered[filtered["volume"].fillna(0) >= min_volume].copy()
    if "openInterest" in filtered.columns:
        filtered = filtered[filtered["openInterest"].fillna(0) >= min_oi].copy()

    return filtered

def add_iv_columns(df, option_label, spot, T, r, price_source):
    df = df.copy()
    df["marketPriceUsed"] = df.apply(
        lambda row: get_market_price_from_row(row, price_source)[0],
        axis=1
    )

    if "impliedVolatility" in df.columns:
        df["iv_yahoo"] = df["impliedVolatility"]
    else:
        df["iv_yahoo"] = np.nan

    if option_label == "Call":
        df["iv_home"] = df.apply(
            lambda row: implied_volatility_call(
                row["marketPriceUsed"], spot, row["strike"], T, r
            ),
            axis=1
        )
    else:
        df["iv_home"] = df.apply(
            lambda row: implied_volatility_put(
                row["marketPriceUsed"], spot, row["strike"], T, r
            ),
            axis=1
        )

    for col in ["iv_yahoo", "iv_home"]:
        df.loc[(df[col] <= 0) | (df[col] > 5), col] = np.nan

    return df

def build_interpolated_surface(df, value_col, min_points_per_row=3, min_points_per_col=2):
    """
    Construit une surface strike x maturité plus lisible
    en interpolant les petits trous de grille.
    """
    surface_pivot = df.pivot_table(
        index="days_to_expiry",
        columns="strike",
        values=value_col,
        aggfunc="mean"
    ).sort_index()

    valid_rows = surface_pivot.notna().sum(axis=1) >= min_points_per_row
    valid_cols = surface_pivot.notna().sum(axis=0) >= min_points_per_col
    surface_pivot = surface_pivot.loc[valid_rows, valid_cols]

    if surface_pivot.empty or surface_pivot.shape[0] < 2 or surface_pivot.shape[1] < 2:
        return None, None, None

    surface_interp = surface_pivot.copy()
    surface_interp = surface_interp.interpolate(axis=1, limit_direction="both")
    surface_interp = surface_interp.interpolate(axis=0, limit_direction="both")

    surface_interp = surface_interp.dropna(axis=0, how="all").dropna(axis=1, how="all")

    if surface_interp.empty or surface_interp.shape[0] < 2 or surface_interp.shape[1] < 2:
        return None, None, None

    x_vals = surface_interp.columns.to_numpy(dtype=float)
    y_vals = surface_interp.index.to_numpy(dtype=float)
    z_vals = surface_interp.to_numpy(dtype=float)

    return x_vals, y_vals, z_vals

# ------------------------------------------------------------
# Volatilité implicite maison
# ------------------------------------------------------------
def implied_volatility_call(market_price, S, K, T, r, sigma_min=1e-6, sigma_max=5.0):
    market_price = float(market_price)

    if market_price <= 0:
        return np.nan

    def objective(sigma):
        return bs_call_price(S, K, T, r, sigma) - market_price

    try:
        return brentq(objective, sigma_min, sigma_max)
    except ValueError:
        return np.nan

def implied_volatility_put(market_price, S, K, T, r, sigma_min=1e-6, sigma_max=5.0):
    market_price = float(market_price)

    if market_price <= 0:
        return np.nan

    def objective(sigma):
        return bs_put_price(S, K, T, r, sigma) - market_price

    try:
        return brentq(objective, sigma_min, sigma_max)
    except ValueError:
        return np.nan

def compute_home_iv(option_label, market_price, S, K, T, r):
    if pd.isna(market_price) or market_price <= 0:
        return np.nan

    try:
        if option_label == "Call":
            return implied_volatility_call(market_price, S, K, T, r)
        return implied_volatility_put(market_price, S, K, T, r)
    except Exception:
        return np.nan

# ------------------------------------------------------------
# Chargement du sous-jacent
# ------------------------------------------------------------
def load_underlying(button=None):
    with market_output:
        market_output.clear_output(wait=True)

        symbol = market_ticker_widget.value.strip().upper()

        try:
            ticker = yf.Ticker(symbol)
            expirations = list(ticker.options)

            if len(expirations) == 0:
                raise ValueError("Aucune échéance d'options disponible pour ce ticker.")

            hist = ticker.history(period="5d")
            if hist.empty:
                raise ValueError("Impossible de récupérer le spot du sous-jacent.")

            spot = float(hist["Close"].dropna().iloc[-1])

            market_state["symbol"] = symbol
            market_state["ticker"] = ticker
            market_state["spot"] = spot
            market_state["expirations"] = expirations
            market_state["calls"] = None
            market_state["puts"] = None
            market_state["expiry"] = None
            market_state["chain_cache"] = {}

            expiry_widget.options = expirations
            expiry_widget.value = expirations[0]

            update_market_panel()

        except Exception as e:
            display(HTML(f"<b style='color:#d62828;'>Erreur chargement sous-jacent :</b> {e}"))

# ------------------------------------------------------------
# Mise à jour du panneau principal
# ------------------------------------------------------------
def update_market_panel(*args):
    if market_state["is_updating"]:
        return

    market_state["is_updating"] = True

    with market_output:
        market_output.clear_output(wait=True)

        try:
            if market_state["ticker"] is None:
                display(HTML("<span style='color:#555;'>Charge d'abord un sous-jacent.</span>"))
                return

            ticker = market_state["ticker"]
            symbol = market_state["symbol"]
            spot = market_state["spot"]
            expiry = expiry_widget.value
            side = option_side_widget.value
            requested_rows = rows_widget.value
            price_source = price_source_widget.value
            max_expiries_to_plot = int(term_expiries_widget.value)
            max_surface_expiries = int(surface_expiries_widget.value)
            n_compare_expiries = int(compare_expiries_widget.value)
            surface_metric = surface_metric_widget.value
            strike_window = float(strike_window_widget.value)
            min_volume = int(min_volume_widget.value)
            min_oi = int(min_oi_widget.value)
            mispricing_top_n = int(mispricing_top_widget.value)

            strike_min = (1 - strike_window) * spot
            strike_max = (1 + strike_window) * spot

            calls_df, puts_df = get_chain_for_expiry(ticker, expiry)

            market_state["calls"] = calls_df
            market_state["puts"] = puts_df
            market_state["expiry"] = expiry

            if side == "calls":
                raw_df = calls_df.copy()
                title_side = "Calls"
                option_label = "Call"
            else:
                raw_df = puts_df.copy()
                title_side = "Puts"
                option_label = "Put"

            df = preprocess_chain_side(raw_df, spot)

            total_available = len(df)
            displayed_rows = min(requested_rows, total_available)
            display_df = df.head(requested_rows).reset_index(drop=True)

            refresh_strike_choices(processed_df=df)

            strike_selected = contract_strike_widget.value
            contract_info_html = ""

            # --------------------------------------------------------
            # Analyse du contrat sélectionné
            # --------------------------------------------------------
            if strike_selected is not None and "strike" in df.columns:
                row_df = df[df["strike"] == strike_selected].copy()

                if not row_df.empty:
                    row = row_df.iloc[0]

                    expiry_date = pd.Timestamp(expiry)
                    today = pd.Timestamp.today().normalize()
                    T = max((expiry_date - today).days / 365.0, 1 / 365.0)

                    r, sigma_model = get_model_params()

                    bid = row.get("bid", np.nan)
                    ask = row.get("ask", np.nan)
                    mid_price = row.get("midPrice", np.nan)
                    last_price = row.get("lastPrice", np.nan)
                    market_iv = row.get("impliedVolatility", np.nan)
                    volume = row.get("volume", np.nan)
                    open_interest = row.get("openInterest", np.nan)
                    itm = row.get("inTheMoney", np.nan)
                    contract_symbol = row.get("contractSymbol", "")

                    market_price, actual_price_source = get_market_price_from_row(row, price_source)

                    if option_label == "Call":
                        bs_price = float(bs_call_price(spot, strike_selected, T, r, sigma_model))
                    else:
                        bs_price = float(bs_put_price(spot, strike_selected, T, r, sigma_model))

                    iv_home = compute_home_iv(option_label, market_price, spot, strike_selected, T, r)
                    price_diff = np.nan if pd.isna(market_price) else float(market_price) - bs_price

                    contract_info_html = f"""
                    <div style="font-family: Arial, sans-serif; margin: 18px 0 10px 0;">
                        <h4 style="margin: 0 0 10px 0;">Contrat sélectionné : {option_label} {strike_selected:.2f}</h4>

                        <div style="display:flex; gap:12px; flex-wrap:wrap; margin-bottom:14px;">
                            <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                                <div style="font-size:12px; color:#6b7280;">Prix de marché</div>
                                <div style="font-size:18px; font-weight:700; margin-top:4px;">{fmt_num(market_price)}</div>
                            </div>
                            <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:200px;">
                                <div style="font-size:12px; color:#6b7280;">Prix Black-Scholes</div>
                                <div style="font-size:18px; font-weight:700; margin-top:4px;">{fmt_num(bs_price)}</div>
                            </div>
                            <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                                <div style="font-size:12px; color:#6b7280;">Écart marché - modèle</div>
                                <div style="font-size:18px; font-weight:700; margin-top:4px;">{fmt_num(price_diff)}</div>
                            </div>
                            <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                                <div style="font-size:12px; color:#6b7280;">Source de prix utilisée</div>
                                <div style="font-size:18px; font-weight:700; margin-top:4px;">{actual_price_source}</div>
                            </div>
                            <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                                <div style="font-size:12px; color:#6b7280;">IV Yahoo</div>
                                <div style="font-size:18px; font-weight:700; margin-top:4px;">{fmt_num(market_iv, pct=True)}</div>
                            </div>
                            <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                                <div style="font-size:12px; color:#6b7280;">IV maison</div>
                                <div style="font-size:18px; font-weight:700; margin-top:4px;">{fmt_num(iv_home, pct=True)}</div>
                            </div>
                        </div>

                        <table style="border-collapse: collapse; width: 100%; background: white; border-radius: 12px; overflow: hidden;">
                            <thead>
                                <tr style="background: #eef2f7;">
                                    <th style="text-align: left; padding: 12px; border-bottom: 1px solid #d9e2f1;">Champ</th>
                                    <th style="text-align: right; padding: 12px; border-bottom: 1px solid #d9e2f1;">Valeur</th>
                                </tr>
                            </thead>
                            <tbody>
                                <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Contract symbol</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{contract_symbol}</td></tr>
                                <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Sous-jacent spot</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{spot:.6f}</td></tr>
                                <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Strike</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{strike_selected:.6f}</td></tr>
                                <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Maturité sélectionnée</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{expiry}</td></tr>
                                <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Temps à maturité (années)</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{T:.6f}</td></tr>
                                <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Taux utilisé</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{r:.6%}</td></tr>
                                <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Vol utilisée dans le modèle</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{sigma_model:.6%}</td></tr>
                                <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">IV Yahoo</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{fmt_num(market_iv, pct=True)}</td></tr>
                                <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">IV maison</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{fmt_num(iv_home, pct=True)}</td></tr>
                                <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Bid</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{fmt_num(bid)}</td></tr>
                                <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Ask</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{fmt_num(ask)}</td></tr>
                                <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Mid price</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{fmt_num(mid_price)}</td></tr>
                                <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Last price</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{fmt_num(last_price)}</td></tr>
                                <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Volume</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{fmt_int(volume)}</td></tr>
                                <tr><td style="padding: 10px 12px; border-bottom: 1px solid #edf1f7;">Open interest</td><td style="padding: 10px 12px; text-align: right; border-bottom: 1px solid #edf1f7;">{fmt_int(open_interest)}</td></tr>
                                <tr><td style="padding: 10px 12px;">In the money</td><td style="padding: 10px 12px; text-align: right;">{itm}</td></tr>
                            </tbody>
                        </table>
                    </div>
                    """

            info_html = f"""
            <div style="font-family: Arial, sans-serif; margin: 8px 0 14px 0;">
                <div style="display:flex; gap:12px; flex-wrap:wrap;">
                    <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                        <div style="font-size:12px; color:#6b7280;">Sous-jacent</div>
                        <div style="font-size:18px; font-weight:700; margin-top:4px;">{symbol}</div>
                    </div>
                    <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                        <div style="font-size:12px; color:#6b7280;">Spot actuel</div>
                        <div style="font-size:18px; font-weight:700; margin-top:4px;">{spot:.2f}</div>
                    </div>
                    <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                        <div style="font-size:12px; color:#6b7280;">Échéance étudiée</div>
                        <div style="font-size:18px; font-weight:700; margin-top:4px;">{expiry}</div>
                    </div>
                    <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:120px;">
                        <div style="font-size:12px; color:#6b7280;">Calls</div>
                        <div style="font-size:18px; font-weight:700; margin-top:4px;">{len(calls_df)}</div>
                    </div>
                    <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:120px;">
                        <div style="font-size:12px; color:#6b7280;">Puts</div>
                        <div style="font-size:18px; font-weight:700; margin-top:4px;">{len(puts_df)}</div>
                    </div>
                </div>
            </div>
            """

            display(HTML(info_html))
            display(HTML(f"<h4 style='margin: 8px 0 10px 0;'>{title_side} les plus proches du spot</h4>"))
            display(HTML(
                f"""
                <p style='margin: 4px 0 10px 0; color:#555; font-family:Arial, sans-serif;'>
                    {displayed_rows} ligne(s) affichée(s) sur {total_available} disponible(s)
                    ({requested_rows} demandée(s)).
                </p>
                """
            ))
            display(HTML(render_dataframe_html(display_df)))

            if contract_info_html:
                display(HTML(contract_info_html))

            # --------------------------------------------------------
            # Smile de volatilité implicite : Yahoo vs IV maison
            # --------------------------------------------------------
            expiry_date = pd.Timestamp(expiry)
            today = pd.Timestamp.today().normalize()
            T_smile = max((expiry_date - today).days / 365.0, 1 / 365.0)
            r_smile, sigma_model_smile = get_model_params()

            df_smile = add_iv_columns(df, option_label, spot, T_smile, r_smile, price_source)
            df_smile["moneyness"] = df_smile["strike"] / spot
            df_smile = df_smile[
                df_smile["marketPriceUsed"].notna() &
                (df_smile["marketPriceUsed"] > 0)
            ].copy()
            df_smile = df_smile.sort_values("strike").reset_index(drop=True)

            smile_info_html = f"""
            <div style="display:flex; gap:12px; flex-wrap:wrap; margin:18px 0 14px 0; font-family:Arial, sans-serif;">
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                    <div style="font-size:12px; color:#6b7280;">Smile affiché</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{title_side}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                    <div style="font-size:12px; color:#6b7280;">Spot actuel</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{spot:.2f}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                    <div style="font-size:12px; color:#6b7280;">Échéance</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{expiry}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                    <div style="font-size:12px; color:#6b7280;">Prix marché utilisé</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{price_source}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                    <div style="font-size:12px; color:#6b7280;">Contrats exploitables</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{len(df_smile)}</div>
                </div>
            </div>
            """
            display(HTML(smile_info_html))

            fig_smile = go.Figure()

            if df_smile["iv_yahoo"].notna().any():
                fig_smile.add_trace(go.Scatter(
                    x=df_smile["strike"],
                    y=df_smile["iv_yahoo"],
                    mode="lines+markers",
                    name="IV Yahoo",
                    hovertemplate="Strike: %{x:.2f}<br>IV Yahoo: %{y:.2%}<extra></extra>"
                ))

            if df_smile["iv_home"].notna().any():
                fig_smile.add_trace(go.Scatter(
                    x=df_smile["strike"],
                    y=df_smile["iv_home"],
                    mode="lines+markers",
                    name="IV maison",
                    hovertemplate="Strike: %{x:.2f}<br>IV maison: %{y:.2%}<extra></extra>"
                ))

            fig_smile.add_vline(
                x=spot,
                line_width=1.2,
                line_dash="dot",
                line_color="#00d4ff"
            )

            fig_smile.update_layout(
                title=f"Smile de volatilité implicite — {title_side} ({symbol})",
                xaxis_title="Strike",
                yaxis_title="Volatilité implicite",
                template="plotly_dark",
                hovermode="x unified",
                height=600,
                legend=dict(
                    orientation="h",
                    yanchor="bottom",
                    y=1.02,
                    xanchor="right",
                    x=1
                )
            )

            fig_smile.update_yaxes(tickformat=".0%")
            fig_smile.show()

            table_cols = ["strike", "marketPriceUsed", "iv_yahoo", "iv_home", "volume", "openInterest", "inTheMoney"]
            table_cols = [c for c in table_cols if c in df_smile.columns]

            display(HTML("<h4 style='margin: 16px 0 8px 0;'>Tableau synthétique du smile</h4>"))
            smile_table_df = df_smile[table_cols].head(20).reset_index(drop=True)
            display(HTML(render_dataframe_html(smile_table_df)))

            # --------------------------------------------------------
            # Structure par terme de la volatilité implicite
            # --------------------------------------------------------
            term_rows = []
            available_expiries = list(market_state["expirations"])[:max_expiries_to_plot]

            for expiry_term in available_expiries:
                try:
                    calls_term, puts_term = get_chain_for_expiry(ticker, expiry_term)

                    if side == "calls":
                        raw_term_df = calls_term.copy()
                        term_option_label = "Call"
                    else:
                        raw_term_df = puts_term.copy()
                        term_option_label = "Put"

                    term_df_local = preprocess_chain_side(raw_term_df, spot).copy()

                    if term_df_local.empty or "strike" not in term_df_local.columns:
                        continue

                    atm_idx = (term_df_local["strike"] - spot).abs().idxmin()
                    atm_row = term_df_local.loc[atm_idx]

                    expiry_date_term = pd.Timestamp(expiry_term)
                    today_term = pd.Timestamp.today().normalize()
                    T_term = max((expiry_date_term - today_term).days / 365.0, 1 / 365.0)
                    days_to_expiry = max((expiry_date_term - today_term).days, 1)

                    market_price_term, actual_price_source_term = get_market_price_from_row(atm_row, price_source)
                    market_iv_term = atm_row.get("impliedVolatility", np.nan)

                    r_term, _ = get_model_params()

                    if term_option_label == "Call":
                        iv_home_term = implied_volatility_call(
                            market_price_term,
                            spot,
                            float(atm_row["strike"]),
                            T_term,
                            r_term
                        )
                    else:
                        iv_home_term = implied_volatility_put(
                            market_price_term,
                            spot,
                            float(atm_row["strike"]),
                            T_term,
                            r_term
                        )

                    if pd.notna(iv_home_term) and (iv_home_term <= 0 or iv_home_term > 5):
                        iv_home_term = np.nan
                    if pd.notna(market_iv_term) and (market_iv_term <= 0 or market_iv_term > 5):
                        market_iv_term = np.nan

                    term_rows.append({
                        "expiry": expiry_term,
                        "days_to_expiry": days_to_expiry,
                        "T_years": T_term,
                        "atm_strike_proxy": float(atm_row["strike"]),
                        "marketPriceUsed": market_price_term,
                        "price_source_used": actual_price_source_term,
                        "iv_yahoo": market_iv_term,
                        "iv_home": iv_home_term,
                        "volume": atm_row.get("volume", np.nan),
                        "openInterest": atm_row.get("openInterest", np.nan)
                    })

                except Exception:
                    continue

            term_df = pd.DataFrame(term_rows)

            if not term_df.empty:
                term_df = term_df.sort_values("days_to_expiry").reset_index(drop=True)

                term_info_html = f"""
                <div style="display:flex; gap:12px; flex-wrap:wrap; margin:18px 0 14px 0; font-family:Arial, sans-serif;">
                    <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                        <div style="font-size:12px; color:#6b7280;">Analyse</div>
                        <div style="font-size:18px; font-weight:700; margin-top:4px;">Structure par terme</div>
                    </div>
                    <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                        <div style="font-size:12px; color:#6b7280;">Chaîne utilisée</div>
                        <div style="font-size:18px; font-weight:700; margin-top:4px;">{title_side}</div>
                    </div>
                    <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                        <div style="font-size:12px; color:#6b7280;">Échéances tracées</div>
                        <div style="font-size:18px; font-weight:700; margin-top:4px;">{len(term_df)}</div>
                    </div>
                    <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                        <div style="font-size:12px; color:#6b7280;">Prix marché utilisé</div>
                        <div style="font-size:18px; font-weight:700; margin-top:4px;">{price_source}</div>
                    </div>
                    <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                        <div style="font-size:12px; color:#6b7280;">Max maturités</div>
                        <div style="font-size:18px; font-weight:700; margin-top:4px;">{max_expiries_to_plot}</div>
                    </div>
                </div>
                """
                display(HTML(term_info_html))

                fig_term = go.Figure()

                if term_df["iv_yahoo"].notna().any():
                    fig_term.add_trace(go.Scatter(
                        x=term_df["days_to_expiry"],
                        y=term_df["iv_yahoo"],
                        mode="lines+markers",
                        name="IV Yahoo",
                        hovertemplate="Jours: %{x}<br>IV Yahoo: %{y:.2%}<extra></extra>"
                    ))

                if term_df["iv_home"].notna().any():
                    fig_term.add_trace(go.Scatter(
                        x=term_df["days_to_expiry"],
                        y=term_df["iv_home"],
                        mode="lines+markers",
                        name="IV maison",
                        hovertemplate="Jours: %{x}<br>IV maison: %{y:.2%}<extra></extra>"
                    ))

                fig_term.update_layout(
                    title=f"Structure par terme de la volatilité implicite — {title_side} ({symbol})",
                    xaxis_title="Jours à maturité",
                    yaxis_title="Volatilité implicite",
                    template="plotly_dark",
                    hovermode="x unified",
                    height=550,
                    legend=dict(
                        orientation="h",
                        yanchor="bottom",
                        y=1.02,
                        xanchor="right",
                        x=1
                    )
                )

                fig_term.update_yaxes(tickformat=".0%")
                fig_term.show()

                term_table_cols = [
                    "expiry",
                    "days_to_expiry",
                    "atm_strike_proxy",
                    "marketPriceUsed",
                    "iv_yahoo",
                    "iv_home",
                    "volume",
                    "openInterest"
                ]
                term_table_cols = [c for c in term_table_cols if c in term_df.columns]

                display(HTML("<h4 style='margin: 16px 0 8px 0;'>Tableau de structure par terme</h4>"))
                display(HTML(render_dataframe_html(term_df[term_table_cols].reset_index(drop=True))))

            # --------------------------------------------------------
            # Dataset filtré pour visualisations avancées
            # --------------------------------------------------------
            advanced_rows = []
            chosen_expiries = list(market_state["expirations"])[:max_surface_expiries]

            for expiry_adv in chosen_expiries:
                try:
                    calls_adv, puts_adv = get_chain_for_expiry(ticker, expiry_adv)

                    if side == "calls":
                        raw_adv_df = calls_adv.copy()
                        adv_option_label = "Call"
                    else:
                        raw_adv_df = puts_adv.copy()
                        adv_option_label = "Put"

                    adv_df_local = preprocess_chain_side(raw_adv_df, spot).copy()
                    adv_df_local = filter_liquid_contracts(
                        adv_df_local,
                        strike_min,
                        strike_max,
                        min_volume,
                        min_oi
                    )

                    if adv_df_local.empty:
                        continue

                    expiry_date_adv = pd.Timestamp(expiry_adv)
                    today_adv = pd.Timestamp.today().normalize()
                    days_adv = max((expiry_date_adv - today_adv).days, 1)
                    T_adv = max(days_adv / 365.0, 1 / 365.0)

                    r_adv, _ = get_model_params()

                    adv_df_local = add_iv_columns(
                        adv_df_local,
                        adv_option_label,
                        spot,
                        T_adv,
                        r_adv,
                        price_source
                    )

                    adv_df_local["days_to_expiry"] = days_adv
                    adv_df_local["expiry"] = expiry_adv

                    advanced_rows.append(adv_df_local)

                except Exception:
                    continue

            advanced_df = pd.DataFrame()

            if len(advanced_rows) > 0:
                advanced_df = pd.concat(advanced_rows, ignore_index=True)

            # --------------------------------------------------------
            # Surface 3D améliorée et interpolée
            # --------------------------------------------------------
            if not advanced_df.empty:
                metric_label = "IV Yahoo" if surface_metric == "iv_yahoo" else "IV maison"
                surface_df = advanced_df[advanced_df[surface_metric].notna()].copy()

                if not surface_df.empty:
                    surface_info_html = f"""
                    <div style="display:flex; gap:12px; flex-wrap:wrap; margin:18px 0 14px 0; font-family:Arial, sans-serif;">
                        <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                            <div style="font-size:12px; color:#6b7280;">Analyse</div>
                            <div style="font-size:18px; font-weight:700; margin-top:4px;">Surface 3D</div>
                        </div>
                        <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                            <div style="font-size:12px; color:#6b7280;">Métrique</div>
                            <div style="font-size:18px; font-weight:700; margin-top:4px;">{metric_label}</div>
                        </div>
                        <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                            <div style="font-size:12px; color:#6b7280;">Fenêtre de strikes</div>
                            <div style="font-size:18px; font-weight:700; margin-top:4px;">[{strike_min:.0f} ; {strike_max:.0f}]</div>
                        </div>
                        <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                            <div style="font-size:12px; color:#6b7280;">Vol min</div>
                            <div style="font-size:18px; font-weight:700; margin-top:4px;">{min_volume}</div>
                        </div>
                        <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                            <div style="font-size:12px; color:#6b7280;">OI min</div>
                            <div style="font-size:18px; font-weight:700; margin-top:4px;">{min_oi}</div>
                        </div>
                    </div>
                    """
                    display(HTML(surface_info_html))

                    x_vals, y_vals, z_vals = build_interpolated_surface(surface_df, surface_metric)

                    if z_vals is not None:
                        zmin = np.nanpercentile(z_vals, 5)
                        zmax = np.nanpercentile(z_vals, 95)
                        z_plot = np.clip(z_vals, zmin, zmax)

                        fig_surface = go.Figure()

                        fig_surface.add_trace(go.Surface(
                            x=x_vals,
                            y=y_vals,
                            z=z_plot,
                            colorscale="Viridis",
                            colorbar=dict(title=metric_label),
                            hovertemplate=(
                                "Strike: %{x:.2f}<br>"
                                "Jours à maturité: %{y:.0f}<br>"
                                f"{metric_label}: " + "%{z:.2%}<extra></extra>"
                            ),
                            showscale=True
                        ))

                        fig_surface.update_layout(
                            title=f"Surface 3D interpolée de volatilité — {metric_label} ({title_side}, {symbol})",
                            template="plotly_dark",
                            height=720,
                            scene=dict(
                                xaxis_title="Strike",
                                yaxis_title="Jours à maturité",
                                zaxis_title=metric_label,
                                camera=dict(eye=dict(x=1.5, y=1.35, z=0.95))
                            ),
                            margin=dict(l=20, r=20, t=60, b=20)
                        )

                        fig_surface.show()

                        display(HTML(
                            "<p style='margin:8px 0 0 0; color:#555; font-family:Arial, sans-serif;'>"
                            "Surface lissée par interpolation des petits trous de grille. "
                            "Elle est plus lisible, mais légèrement moins brute que les données originales."
                            "</p>"
                        ))

            # --------------------------------------------------------
            # Smiles multi-maturités : visuel le plus lisible
            # --------------------------------------------------------
            if not advanced_df.empty:
                metric_label = "IV Yahoo" if surface_metric == "iv_yahoo" else "IV maison"
                multi_df = advanced_df[advanced_df[surface_metric].notna()].copy()

                if not multi_df.empty:
                    unique_expiries = (
                        multi_df[["expiry", "days_to_expiry"]]
                        .drop_duplicates()
                        .sort_values("days_to_expiry")
                        .head(n_compare_expiries)
                    )

                    display(HTML("<h4 style='margin: 18px 0 8px 0;'>Smiles comparés sur plusieurs maturités</h4>"))

                    fig_multi_smile = go.Figure()

                    for _, exp_row in unique_expiries.iterrows():
                        expiry_label = exp_row["expiry"]
                        days_label = int(exp_row["days_to_expiry"])

                        tmp = multi_df[multi_df["expiry"] == expiry_label].copy()
                        tmp = tmp.sort_values("strike")

                        if tmp.empty:
                            continue

                        fig_multi_smile.add_trace(go.Scatter(
                            x=tmp["strike"],
                            y=tmp[surface_metric],
                            mode="lines+markers",
                            name=f"{expiry_label} ({days_label}j)",
                            hovertemplate=(
                                "Strike: %{x:.2f}<br>"
                                f"{metric_label}: " + "%{y:.2%}<extra></extra>"
                            )
                        ))

                    fig_multi_smile.add_vline(
                        x=spot,
                        line_width=1.2,
                        line_dash="dot",
                        line_color="#00d4ff"
                    )

                    fig_multi_smile.update_layout(
                        title=f"Comparaison de smiles — {metric_label} ({title_side}, {symbol})",
                        xaxis_title="Strike",
                        yaxis_title=metric_label,
                        template="plotly_dark",
                        hovermode="x unified",
                        height=650,
                        legend=dict(
                            orientation="h",
                            yanchor="bottom",
                            y=1.02,
                            xanchor="right",
                            x=1
                        )
                    )

                    fig_multi_smile.update_yaxes(tickformat=".0%")
                    fig_multi_smile.show()

            # --------------------------------------------------------
            # Mispricing marché vs modèle sur l'échéance sélectionnée
            # --------------------------------------------------------
            if not df.empty and "strike" in df.columns:
                df_mispricing = df.copy()
                expiry_date_mis = pd.Timestamp(expiry)
                today_mis = pd.Timestamp.today().normalize()
                T_mis = max((expiry_date_mis - today_mis).days / 365.0, 1 / 365.0)

                r_mis, sigma_model_mis = get_model_params()

                df_mispricing["marketPriceUsed"] = df_mispricing.apply(
                    lambda row: get_market_price_from_row(row, price_source)[0],
                    axis=1
                )

                if option_label == "Call":
                    df_mispricing["bs_model_price"] = bs_call_price(
                        spot,
                        df_mispricing["strike"].to_numpy(dtype=float),
                        T_mis,
                        r_mis,
                        sigma_model_mis
                    )
                    df_mispricing["iv_home"] = df_mispricing.apply(
                        lambda row: implied_volatility_call(
                            row["marketPriceUsed"], spot, row["strike"], T_mis, r_mis
                        ),
                        axis=1
                    )
                else:
                    df_mispricing["bs_model_price"] = bs_put_price(
                        spot,
                        df_mispricing["strike"].to_numpy(dtype=float),
                        T_mis,
                        r_mis,
                        sigma_model_mis
                    )
                    df_mispricing["iv_home"] = df_mispricing.apply(
                        lambda row: implied_volatility_put(
                            row["marketPriceUsed"], spot, row["strike"], T_mis, r_mis
                        ),
                        axis=1
                    )

                if "impliedVolatility" in df_mispricing.columns:
                    df_mispricing["iv_yahoo"] = df_mispricing["impliedVolatility"]
                else:
                    df_mispricing["iv_yahoo"] = np.nan

                df_mispricing["price_diff_model"] = df_mispricing["marketPriceUsed"] - df_mispricing["bs_model_price"]
                df_mispricing["abs_diff_model"] = df_mispricing["price_diff_model"].abs()
                df_mispricing["moneyness"] = df_mispricing["strike"].apply(
                    lambda k: classify_moneyness(option_label, spot, k)
                )

                df_mispricing = df_mispricing[
                    df_mispricing["marketPriceUsed"].notna() &
                    (df_mispricing["marketPriceUsed"] > 0)
                ].copy()

                df_mispricing = df_mispricing.sort_values("abs_diff_model", ascending=False).reset_index(drop=True)

                if not df_mispricing.empty:
                    mis_info_html = f"""
                    <div style="display:flex; gap:12px; flex-wrap:wrap; margin:18px 0 14px 0; font-family:Arial, sans-serif;">
                        <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                            <div style="font-size:12px; color:#6b7280;">Analyse</div>
                            <div style="font-size:18px; font-weight:700; margin-top:4px;">Mispricing</div>
                        </div>
                        <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                            <div style="font-size:12px; color:#6b7280;">Chaîne</div>
                            <div style="font-size:18px; font-weight:700; margin-top:4px;">{title_side}</div>
                        </div>
                        <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                            <div style="font-size:12px; color:#6b7280;">Échéance</div>
                            <div style="font-size:18px; font-weight:700; margin-top:4px;">{expiry}</div>
                        </div>
                        <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                            <div style="font-size:12px; color:#6b7280;">Vol modèle</div>
                            <div style="font-size:18px; font-weight:700; margin-top:4px;">{sigma_model_mis:.2%}</div>
                        </div>
                        <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                            <div style="font-size:12px; color:#6b7280;">Top affichés</div>
                            <div style="font-size:18px; font-weight:700; margin-top:4px;">{mispricing_top_n}</div>
                        </div>
                    </div>
                    """
                    display(HTML(mis_info_html))

                    mis_cols = [
                        "strike",
                        "moneyness",
                        "marketPriceUsed",
                        "bs_model_price",
                        "price_diff_model",
                        "abs_diff_model",
                        "iv_yahoo",
                        "iv_home",
                        "volume",
                        "openInterest",
                        "inTheMoney"
                    ]
                    mis_cols = [c for c in mis_cols if c in df_mispricing.columns]

                    display(HTML("<h4 style='margin: 16px 0 8px 0;'>Tableau des plus grands écarts marché vs modèle</h4>"))
                    mispricing_table_df = df_mispricing[mis_cols].head(mispricing_top_n).reset_index(drop=True)
                    display(HTML(render_dataframe_html(mispricing_table_df)))

        except Exception as e:
            display(HTML(f"<b style='color:#d62828;'>Erreur panneau marché :</b> {e}"))

        finally:
            market_state["is_updating"] = False

load_market_button.on_click(load_underlying)

observed_widgets = [
    expiry_widget,
    option_side_widget,
    rows_widget,
    contract_strike_widget,
    price_source_widget,
    term_expiries_widget,
    surface_expiries_widget,
    compare_expiries_widget,
    surface_metric_widget,
    strike_window_widget,
    min_volume_widget,
    min_oi_widget,
    mispricing_top_widget
]

if "r_widget" in globals():
    observed_widgets.append(r_widget)
if "sigma_widget" in globals():
    observed_widgets.append(sigma_widget)

for w in observed_widgets:
    w.observe(update_market_panel, names="value")

market_ui = widgets.VBox([
    widgets.HTML("<h3 style='margin:0;'>Panneau marché : sous-jacent, chaîne, contrat, smile, term structure, surface 3D et mispricing</h3>"),
    widgets.HTML("<p style='margin:6px 0 10px 0; color:#555;'>Entre un ticker Yahoo Finance, charge le sous-jacent, puis explore automatiquement les calls et puts par échéance. Tu peux sélectionner un strike précis, comparer prix marché vs prix Black-Scholes, calculer l’IV maison, visualiser le smile, la structure par terme, une surface 3D filtrée et interpolée, et des smiles comparés plus lisibles dans le même panneau.</p>"),
    widgets.HBox([market_ticker_widget, load_market_button, expiry_widget, option_side_widget]),
    widgets.HBox([rows_widget, contract_strike_widget, price_source_widget]),
    widgets.HBox([term_expiries_widget, surface_expiries_widget, compare_expiries_widget]),
    widgets.HBox([surface_metric_widget, strike_window_widget, min_volume_widget, min_oi_widget]),
    widgets.HBox([mispricing_top_widget])
])

display(market_ui, market_output)

# Chargement initial
load_underlying()

# ============================================================
# 3.2 Greeks de marché sur la chaîne réelle
# Comparaison Yahoo vs Maison
# ============================================================

from scipy.stats import norm

greeks_curve_widget = widgets.Dropdown(
    options=[
        ("Delta", "delta"),
        ("Gamma", "gamma"),
        ("Vega (+1 pt vol)", "vega_1pct"),
        ("Theta (/jour)", "theta_1d"),
        ("Rho (+1%)", "rho_1pct")
    ],
    value="delta",
    description="Courbe",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="240px")
)

greeks_rows_widget = widgets.IntSlider(
    value=15,
    min=5,
    max=30,
    step=1,
    description="Lignes",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="260px")
)

greeks_output = widgets.Output()


# ------------------------------------------------------------
# Helpers Greeks
# ------------------------------------------------------------
def _safe_bs_d1(S, K, T, r, sigma):
    return (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))

def _safe_bs_d2(S, K, T, r, sigma):
    return _safe_bs_d1(S, K, T, r, sigma) - sigma * np.sqrt(T)

def add_greeks_with_prefix(df, option_label, S, T, r, iv_col, prefix):
    """
    Calcule les Greeks à partir d'une colonne de vol donnée
    et ajoute les colonnes suffixées par le préfixe demandé.
    """
    out = df.copy()

    delta_col = f"delta_{prefix}"
    gamma_col = f"gamma_{prefix}"
    vega_col = f"vega_1pct_{prefix}"
    theta_col = f"theta_1d_{prefix}"
    rho_col = f"rho_1pct_{prefix}"

    out[delta_col] = np.nan
    out[gamma_col] = np.nan
    out[vega_col] = np.nan
    out[theta_col] = np.nan
    out[rho_col] = np.nan

    if iv_col not in out.columns:
        return out

    K = out["strike"].astype(float).to_numpy()
    sigma = out[iv_col].astype(float).to_numpy()

    valid = (
        np.isfinite(K) &
        np.isfinite(sigma) &
        (K > 0) &
        (sigma > 0) &
        np.isfinite(S) &
        (S > 0) &
        np.isfinite(T) &
        (T > 0)
    )

    if np.any(valid):
        K_v = K[valid]
        sigma_v = sigma[valid]

        d1 = _safe_bs_d1(S, K_v, T, r, sigma_v)
        d2 = _safe_bs_d2(S, K_v, T, r, sigma_v)
        phi_d1 = norm.pdf(d1)

        gamma_raw = phi_d1 / (S * sigma_v * np.sqrt(T))
        vega_raw = S * phi_d1 * np.sqrt(T)

        if option_label == "Call":
            delta_raw = norm.cdf(d1)
            theta_raw = (
                -(S * phi_d1 * sigma_v) / (2 * np.sqrt(T))
                - r * K_v * np.exp(-r * T) * norm.cdf(d2)
            )
            rho_raw = K_v * T * np.exp(-r * T) * norm.cdf(d2)
        else:
            delta_raw = norm.cdf(d1) - 1.0
            theta_raw = (
                -(S * phi_d1 * sigma_v) / (2 * np.sqrt(T))
                + r * K_v * np.exp(-r * T) * norm.cdf(-d2)
            )
            rho_raw = -K_v * T * np.exp(-r * T) * norm.cdf(-d2)

        out.loc[valid, delta_col] = delta_raw
        out.loc[valid, gamma_col] = gamma_raw
        out.loc[valid, vega_col] = vega_raw * 0.01      # pour +1 point de vol
        out.loc[valid, theta_col] = theta_raw / 365.0   # par jour
        out.loc[valid, rho_col] = rho_raw * 0.01        # pour +1% de taux

    return out


# ------------------------------------------------------------
# Mise à jour de la section Greeks
# ------------------------------------------------------------
def update_market_greeks(*args):
    with greeks_output:
        greeks_output.clear_output(wait=True)

        try:
            if market_state["ticker"] is None or market_state["spot"] is None:
                display(HTML("<span style='color:#555;'>Charge d'abord un sous-jacent dans le panneau marché.</span>"))
                return

            ticker = market_state["ticker"]
            symbol = market_state["symbol"]
            spot = float(market_state["spot"])
            expiry = expiry_widget.value
            side = option_side_widget.value
            strike_selected = contract_strike_widget.value
            price_source = price_source_widget.value
            greek_curve = greeks_curve_widget.value
            n_rows = int(greeks_rows_widget.value)

            expiry_date = pd.Timestamp(expiry)
            today = pd.Timestamp.today().normalize()
            T = max((expiry_date - today).days / 365.0, 1 / 365.0)

            r_value, _ = get_model_params()

            calls_df, puts_df = get_chain_for_expiry(ticker, expiry)

            if side == "calls":
                raw_df = calls_df.copy()
                title_side = "Calls"
                option_label = "Call"
            else:
                raw_df = puts_df.copy()
                title_side = "Puts"
                option_label = "Put"

            chain_df = preprocess_chain_side(raw_df, spot).copy()

            strike_window = float(strike_window_widget.value) if "strike_window_widget" in globals() else 0.25
            min_volume = int(min_volume_widget.value) if "min_volume_widget" in globals() else 0
            min_oi = int(min_oi_widget.value) if "min_oi_widget" in globals() else 0

            strike_min = (1 - strike_window) * spot
            strike_max = (1 + strike_window) * spot

            chain_df = filter_liquid_contracts(
                chain_df,
                strike_min=strike_min,
                strike_max=strike_max,
                min_volume=min_volume,
                min_oi=min_oi
            )

            if chain_df.empty:
                display(HTML("<span style='color:#d62828;'>Aucune option disponible après filtrage pour calculer les Greeks.</span>"))
                return

            chain_df = add_iv_columns(
                chain_df,
                option_label=option_label,
                spot=spot,
                T=T,
                r=r_value,
                price_source=price_source
            )

            chain_df = add_greeks_with_prefix(
                chain_df,
                option_label=option_label,
                S=spot,
                T=T,
                r=r_value,
                iv_col="iv_yahoo",
                prefix="yahoo"
            )

            chain_df = add_greeks_with_prefix(
                chain_df,
                option_label=option_label,
                S=spot,
                T=T,
                r=r_value,
                iv_col="iv_home",
                prefix="home"
            )

            chain_df = chain_df.sort_values("strike").reset_index(drop=True)

            info_html = f"""
            <div style="display:flex; gap:12px; flex-wrap:wrap; margin:10px 0 14px 0; font-family:Arial, sans-serif;">
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                    <div style="font-size:12px; color:#6b7280;">Analyse</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">Greeks de marché</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                    <div style="font-size:12px; color:#6b7280;">Sous-jacent</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{symbol}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                    <div style="font-size:12px; color:#6b7280;">Chaîne</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{title_side}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                    <div style="font-size:12px; color:#6b7280;">Échéance</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{expiry}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                    <div style="font-size:12px; color:#6b7280;">Spot</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{spot:.2f}</div>
                </div>
            </div>
            """
            display(HTML(info_html))

            # --------------------------------------------------------
            # Greeks du contrat sélectionné : comparaison Yahoo vs Maison
            # --------------------------------------------------------
            if strike_selected is not None and "strike" in chain_df.columns:
                selected_df = chain_df[chain_df["strike"] == strike_selected].copy()

                if not selected_df.empty:
                    row = selected_df.iloc[0]

                    selected_html = f"""
                    <div style="font-family: Arial, sans-serif; margin: 8px 0 12px 0;">
                        <h4 style="margin: 0 0 10px 0;">Greeks du contrat sélectionné : {option_label} {strike_selected:.2f}</h4>

                        <table style="border-collapse: collapse; width: 100%; background: white; border-radius: 12px; overflow: hidden;">
                            <thead>
                                <tr style="background: #eef2f7;">
                                    <th style="text-align: left; padding: 12px; border-bottom: 1px solid #d9e2f1;">Métrique</th>
                                    <th style="text-align: right; padding: 12px; border-bottom: 1px solid #d9e2f1;">Yahoo</th>
                                    <th style="text-align: right; padding: 12px; border-bottom: 1px solid #d9e2f1;">Maison</th>
                                </tr>
                            </thead>
                            <tbody>
                                <tr><td style="padding:10px 12px; border-bottom:1px solid #edf1f7;">IV</td><td style="padding:10px 12px; text-align:right; border-bottom:1px solid #edf1f7;">{fmt_num(row.get('iv_yahoo', np.nan), pct=True)}</td><td style="padding:10px 12px; text-align:right; border-bottom:1px solid #edf1f7;">{fmt_num(row.get('iv_home', np.nan), pct=True)}</td></tr>
                                <tr><td style="padding:10px 12px; border-bottom:1px solid #edf1f7;">Delta</td><td style="padding:10px 12px; text-align:right; border-bottom:1px solid #edf1f7;">{fmt_num(row.get('delta_yahoo', np.nan))}</td><td style="padding:10px 12px; text-align:right; border-bottom:1px solid #edf1f7;">{fmt_num(row.get('delta_home', np.nan))}</td></tr>
                                <tr><td style="padding:10px 12px; border-bottom:1px solid #edf1f7;">Gamma</td><td style="padding:10px 12px; text-align:right; border-bottom:1px solid #edf1f7;">{fmt_num(row.get('gamma_yahoo', np.nan))}</td><td style="padding:10px 12px; text-align:right; border-bottom:1px solid #edf1f7;">{fmt_num(row.get('gamma_home', np.nan))}</td></tr>
                                <tr><td style="padding:10px 12px; border-bottom:1px solid #edf1f7;">Vega (+1 pt vol)</td><td style="padding:10px 12px; text-align:right; border-bottom:1px solid #edf1f7;">{fmt_num(row.get('vega_1pct_yahoo', np.nan))}</td><td style="padding:10px 12px; text-align:right; border-bottom:1px solid #edf1f7;">{fmt_num(row.get('vega_1pct_home', np.nan))}</td></tr>
                                <tr><td style="padding:10px 12px; border-bottom:1px solid #edf1f7;">Theta (/jour)</td><td style="padding:10px 12px; text-align:right; border-bottom:1px solid #edf1f7;">{fmt_num(row.get('theta_1d_yahoo', np.nan))}</td><td style="padding:10px 12px; text-align:right; border-bottom:1px solid #edf1f7;">{fmt_num(row.get('theta_1d_home', np.nan))}</td></tr>
                                <tr><td style="padding:10px 12px;">Rho (+1%)</td><td style="padding:10px 12px; text-align:right;">{fmt_num(row.get('rho_1pct_yahoo', np.nan))}</td><td style="padding:10px 12px; text-align:right;">{fmt_num(row.get('rho_1pct_home', np.nan))}</td></tr>
                            </tbody>
                        </table>
                    </div>
                    """
                    display(HTML(selected_html))

            # --------------------------------------------------------
            # Courbe d'un Greek : Yahoo + Maison ensemble
            # --------------------------------------------------------
            metric_label_map = {
                "delta": "Delta",
                "gamma": "Gamma",
                "vega_1pct": "Vega (+1 pt vol)",
                "theta_1d": "Theta (/jour)",
                "rho_1pct": "Rho (+1%)"
            }
            metric_label = metric_label_map[greek_curve]

            yahoo_col = f"{greek_curve}_yahoo"
            home_col = f"{greek_curve}_home"

            plot_df = chain_df.copy()

            fig_greek = go.Figure()

            if yahoo_col in plot_df.columns and plot_df[yahoo_col].notna().any():
                fig_greek.add_trace(go.Scatter(
                    x=plot_df["strike"],
                    y=plot_df[yahoo_col],
                    mode="lines+markers",
                    name=f"{metric_label} Yahoo",
                    hovertemplate=(
                        "Strike: %{x:.2f}<br>"
                        f"{metric_label} Yahoo: " + "%{y:.6f}<extra></extra>"
                    )
                ))

            if home_col in plot_df.columns and plot_df[home_col].notna().any():
                fig_greek.add_trace(go.Scatter(
                    x=plot_df["strike"],
                    y=plot_df[home_col],
                    mode="lines+markers",
                    name=f"{metric_label} Maison",
                    hovertemplate=(
                        "Strike: %{x:.2f}<br>"
                        f"{metric_label} Maison: " + "%{y:.6f}<extra></extra>"
                    )
                ))

            fig_greek.add_vline(
                x=spot,
                line_width=1.2,
                line_dash="dot",
                line_color="#00d4ff"
            )

            fig_greek.update_layout(
                title=f"{metric_label} selon le strike — comparaison Yahoo vs Maison ({title_side}, {symbol})",
                xaxis_title="Strike",
                yaxis_title=metric_label,
                template="plotly_dark",
                hovermode="x unified",
                height=550
            )

            fig_greek.show()

            # --------------------------------------------------------
            # Tableau complet Yahoo + Maison
            # --------------------------------------------------------
            table_cols = [
                "strike",
                "marketPriceUsed",
                "iv_yahoo",
                "iv_home",
                "delta_yahoo",
                "delta_home",
                "gamma_yahoo",
                "gamma_home",
                "vega_1pct_yahoo",
                "vega_1pct_home",
                "theta_1d_yahoo",
                "theta_1d_home",
                "rho_1pct_yahoo",
                "rho_1pct_home",
                "volume",
                "openInterest",
                "inTheMoney"
            ]
            table_cols = [c for c in table_cols if c in chain_df.columns]

            display(HTML("<h4 style='margin: 16px 0 8px 0;'>Tableau comparatif des Greeks : Yahoo vs Maison</h4>"))
            table_df = chain_df[table_cols].head(n_rows).reset_index(drop=True)
            display(HTML(render_dataframe_html(table_df)))

        except Exception as e:
            display(HTML(f"<b style='color:#d62828;'>Erreur Greeks marché :</b> {e}"))


# ------------------------------------------------------------
# UI
# ------------------------------------------------------------
greeks_ui = widgets.VBox([
    widgets.HTML("<h3 style='margin:16px 0 6px 0;'>Greeks de marché sur la chaîne réelle</h3>"),
    widgets.HTML("<p style='margin:0 0 10px 0; color:#555;'>Cette section compare directement Yahoo et Maison sur les Greeks, en réutilisant le ticker, l’échéance, le side et le strike du panneau marché.</p>"),
    widgets.HBox([greeks_curve_widget, greeks_rows_widget])
])

display(greeks_ui, greeks_output)

# ------------------------------------------------------------
# Synchronisation avec le panneau principal
# ------------------------------------------------------------
greeks_observed_widgets = [
    greeks_curve_widget,
    greeks_rows_widget,
    expiry_widget,
    option_side_widget,
    contract_strike_widget,
    price_source_widget
]

if "strike_window_widget" in globals():
    greeks_observed_widgets.append(strike_window_widget)
if "min_volume_widget" in globals():
    greeks_observed_widgets.append(min_volume_widget)
if "min_oi_widget" in globals():
    greeks_observed_widgets.append(min_oi_widget)
if "r_widget" in globals():
    greeks_observed_widgets.append(r_widget)

for w in greeks_observed_widgets:
    w.observe(update_market_greeks, names="value")

# Premier affichage
update_market_greeks()
```

```python
# ============================================================
# 4. Delta hedging dynamique connecté au panneau marché
# Comparaison directe Yahoo vs Maison
# ============================================================

hedge_steps_widget = widgets.IntSlider(
    value=25,
    min=5,
    max=100,
    step=1,
    description="Pas",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="240px")
)

hedge_qty_widget = widgets.IntSlider(
    value=1,
    min=1,
    max=100,
    step=1,
    description="Nb opt",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="240px")
)

hedge_mu_widget = widgets.FloatSlider(
    value=0.00,
    min=-0.30,
    max=0.30,
    step=0.01,
    description="Drift μ",
    readout_format=".0%",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="260px")
)

hedge_seed_widget = widgets.IntText(
    value=42,
    description="Seed",
    style={"description_width": "80px"},
    layout=widgets.Layout(width="180px")
)

hedge_run_button = widgets.Button(
    description="Lancer la simulation",
    button_style="primary",
    layout=widgets.Layout(width="180px")
)

hedge_output = widgets.Output()


# ------------------------------------------------------------
# Helpers pricing / delta
# ------------------------------------------------------------
def bs_price_by_type(option_label, S, K, T, r, sigma):
    if option_label == "Call":
        return float(bs_call_price(S, K, T, r, sigma))
    return float(bs_put_price(S, K, T, r, sigma))

def bs_delta_by_type(option_label, S, K, T, r, sigma):
    if option_label == "Call":
        return float(call_delta(S, K, T, r, sigma))
    return float(put_delta(S, K, T, r, sigma))

def get_selected_contract_row():
    """
    Récupère la ligne du contrat actuellement sélectionné
    dans le panneau marché du dessus.
    """
    if market_state["ticker"] is None or market_state["spot"] is None:
        return None, None, None, None

    ticker = market_state["ticker"]
    spot = float(market_state["spot"])
    expiry = expiry_widget.value
    side = option_side_widget.value
    strike_selected = contract_strike_widget.value

    calls_df, puts_df = get_chain_for_expiry(ticker, expiry)

    if side == "calls":
        raw_df = calls_df.copy()
        option_label = "Call"
    else:
        raw_df = puts_df.copy()
        option_label = "Put"

    df = preprocess_chain_side(raw_df, spot).copy()

    if df.empty or strike_selected is None:
        return None, None, None, None

    exact = df[df["strike"] == strike_selected].copy()
    if not exact.empty:
        row = exact.iloc[0]
    else:
        idx = (df["strike"] - strike_selected).abs().idxmin()
        row = df.loc[idx]

    return row, option_label, spot, expiry


# ------------------------------------------------------------
# Simulation delta hedging : comparaison Yahoo vs Maison
# ------------------------------------------------------------
def run_delta_hedge_simulation(button=None):
    with hedge_output:
        hedge_output.clear_output(wait=True)

        try:
            row, option_label, spot, expiry = get_selected_contract_row()

            if row is None:
                display(HTML("<span style='color:#d62828;'>Impossible de récupérer le contrat sélectionné depuis le panneau marché.</span>"))
                return

            K = float(row["strike"])
            qty = int(hedge_qty_widget.value)
            n_steps = int(hedge_steps_widget.value)
            mu = float(hedge_mu_widget.value)
            seed = int(hedge_seed_widget.value)

            expiry_date = pd.Timestamp(expiry)
            today = pd.Timestamp.today().normalize()
            T_total = max((expiry_date - today).days / 365.0, 1 / 365.0)

            r_value, _ = get_model_params()

            # Recalcule IV Yahoo / IV maison pour la ligne sélectionnée
            tmp_df = pd.DataFrame([row]).copy()
            tmp_df = add_iv_columns(
                tmp_df,
                option_label=option_label,
                spot=spot,
                T=T_total,
                r=r_value,
                price_source=price_source_widget.value
            )
            tmp_row = tmp_df.iloc[0]

            sigma_yahoo = tmp_row.get("iv_yahoo", np.nan)
            sigma_home = tmp_row.get("iv_home", np.nan)

            # Fallbacks propres
            if pd.isna(sigma_yahoo) or sigma_yahoo <= 0:
                sigma_yahoo = sigma_home
            if pd.isna(sigma_home) or sigma_home <= 0:
                sigma_home = sigma_yahoo
            if pd.isna(sigma_yahoo) or sigma_yahoo <= 0:
                sigma_yahoo = 0.20
            if pd.isna(sigma_home) or sigma_home <= 0:
                sigma_home = 0.20

            sigma_yahoo = float(sigma_yahoo)
            sigma_home = float(sigma_home)

            # Une seule trajectoire commune pour comparer proprement les deux hedges
            valid_sigmas = [x for x in [sigma_yahoo, sigma_home] if np.isfinite(x) and x > 0]
            sigma_path = float(np.mean(valid_sigmas)) if len(valid_sigmas) > 0 else 0.20

            # Conditions initiales
            option_price_y0 = bs_price_by_type(option_label, spot, K, T_total, r_value, sigma_yahoo)
            option_price_h0 = bs_price_by_type(option_label, spot, K, T_total, r_value, sigma_home)

            delta_y0 = bs_delta_by_type(option_label, spot, K, T_total, r_value, sigma_yahoo)
            delta_h0 = bs_delta_by_type(option_label, spot, K, T_total, r_value, sigma_home)

            # Grille de temps
            times = np.linspace(0, T_total, n_steps + 1)
            dt = T_total / n_steps

            # Trajectoire spot commune
            np.random.seed(seed)
            z = np.random.normal(size=n_steps)

            spots = np.zeros(n_steps + 1)

            option_y = np.zeros(n_steps + 1)
            option_h = np.zeros(n_steps + 1)

            delta_y = np.zeros(n_steps + 1)
            delta_h = np.zeros(n_steps + 1)

            cash_y = np.zeros(n_steps + 1)
            cash_h = np.zeros(n_steps + 1)

            hedged_pnl_y = np.zeros(n_steps + 1)
            hedged_pnl_h = np.zeros(n_steps + 1)

            unhedged_pnl_y = np.zeros(n_steps + 1)
            unhedged_pnl_h = np.zeros(n_steps + 1)

            spots[0] = spot
            option_y[0] = option_price_y0
            option_h[0] = option_price_h0
            delta_y[0] = delta_y0
            delta_h[0] = delta_h0

            # Portefeuille hedgé d'une position LONG option
            cash_y[0] = qty * (delta_y0 * spot - option_price_y0)
            cash_h[0] = qty * (delta_h0 * spot - option_price_h0)

            hedged_pnl_y[0] = qty * option_y[0] - qty * delta_y[0] * spot + cash_y[0]
            hedged_pnl_h[0] = qty * option_h[0] - qty * delta_h[0] * spot + cash_h[0]

            for i in range(1, n_steps + 1):
                # Spot commun simulé
                spots[i] = spots[i - 1] * np.exp(
                    (mu - 0.5 * sigma_path**2) * dt + sigma_path * np.sqrt(dt) * z[i - 1]
                )

                tau = max(T_total - times[i], 1e-8)

                # --- Yahoo ---
                cash_before_y = cash_y[i - 1] * np.exp(r_value * dt)
                option_y[i] = bs_price_by_type(option_label, spots[i], K, tau, r_value, sigma_yahoo)
                delta_prev_y = delta_y[i - 1]

                if tau <= 1e-8:
                    if option_label == "Call":
                        delta_new_y = 1.0 if spots[i] > K else 0.0
                    else:
                        delta_new_y = -1.0 if spots[i] < K else 0.0
                else:
                    delta_new_y = bs_delta_by_type(option_label, spots[i], K, tau, r_value, sigma_yahoo)

                delta_y[i] = delta_new_y
                cash_y[i] = cash_before_y + qty * (delta_new_y - delta_prev_y) * spots[i]
                hedged_pnl_y[i] = qty * option_y[i] - qty * delta_y[i] * spots[i] + cash_y[i]
                unhedged_pnl_y[i] = qty * (option_y[i] - option_price_y0)

                # --- Maison ---
                cash_before_h = cash_h[i - 1] * np.exp(r_value * dt)
                option_h[i] = bs_price_by_type(option_label, spots[i], K, tau, r_value, sigma_home)
                delta_prev_h = delta_h[i - 1]

                if tau <= 1e-8:
                    if option_label == "Call":
                        delta_new_h = 1.0 if spots[i] > K else 0.0
                    else:
                        delta_new_h = -1.0 if spots[i] < K else 0.0
                else:
                    delta_new_h = bs_delta_by_type(option_label, spots[i], K, tau, r_value, sigma_home)

                delta_h[i] = delta_new_h
                cash_h[i] = cash_before_h + qty * (delta_new_h - delta_prev_h) * spots[i]
                hedged_pnl_h[i] = qty * option_h[i] - qty * delta_h[i] * spots[i] + cash_h[i]
                unhedged_pnl_h[i] = qty * (option_h[i] - option_price_h0)

            results_df = pd.DataFrame({
                "step": np.arange(n_steps + 1),
                "time_years": times,
                "spot": spots,
                "option_price_yahoo": option_y,
                "option_price_home": option_h,
                "delta_yahoo": delta_y,
                "delta_home": delta_h,
                "hedged_pnl_yahoo": hedged_pnl_y,
                "hedged_pnl_home": hedged_pnl_h,
                "unhedged_pnl_yahoo": unhedged_pnl_y,
                "unhedged_pnl_home": unhedged_pnl_h
            })

            info_html = f"""
            <div style="display:flex; gap:12px; flex-wrap:wrap; margin:10px 0 14px 0; font-family:Arial, sans-serif;">
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                    <div style="font-size:12px; color:#6b7280;">Sous-jacent</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{market_state['symbol']}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                    <div style="font-size:12px; color:#6b7280;">Contrat</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{option_label} {K:.2f}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                    <div style="font-size:12px; color:#6b7280;">Échéance</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{expiry}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                    <div style="font-size:12px; color:#6b7280;">IV Yahoo</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{sigma_yahoo:.2%}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:160px;">
                    <div style="font-size:12px; color:#6b7280;">IV Maison</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{sigma_home:.2%}</div>
                </div>
                <div style="background:#f5f7fb; border:1px solid #d9e2f1; border-radius:10px; padding:10px 14px; min-width:180px;">
                    <div style="font-size:12px; color:#6b7280;">Vol de trajectoire</div>
                    <div style="font-size:18px; font-weight:700; margin-top:4px;">{sigma_path:.2%}</div>
                </div>
            </div>
            """
            display(HTML(info_html))

            # --------------------------------------------------------
            # Graphique spot simulé
            # --------------------------------------------------------
            fig_spot = go.Figure()
            fig_spot.add_trace(go.Scatter(
                x=results_df["step"],
                y=results_df["spot"],
                mode="lines+markers",
                name="Spot simulé",
                hovertemplate="Step: %{x}<br>Spot: %{y:.4f}<extra></extra>"
            ))
            fig_spot.update_layout(
                title=f"Trajectoire simulée du sous-jacent — {market_state['symbol']}",
                xaxis_title="Pas de temps",
                yaxis_title="Spot",
                template="plotly_dark",
                height=420
            )
            fig_spot.show()

            # --------------------------------------------------------
            # Graphique delta : Yahoo vs Maison
            # --------------------------------------------------------
            fig_delta = go.Figure()
            fig_delta.add_trace(go.Scatter(
                x=results_df["step"],
                y=results_df["delta_yahoo"],
                mode="lines+markers",
                name="Delta Yahoo",
                hovertemplate="Step: %{x}<br>Delta Yahoo: %{y:.6f}<extra></extra>"
            ))
            fig_delta.add_trace(go.Scatter(
                x=results_df["step"],
                y=results_df["delta_home"],
                mode="lines+markers",
                name="Delta Maison",
                hovertemplate="Step: %{x}<br>Delta Maison: %{y:.6f}<extra></extra>"
            ))
            fig_delta.update_layout(
                title=f"Delta dynamique — comparaison Yahoo vs Maison ({option_label} {K:.2f})",
                xaxis_title="Pas de temps",
                yaxis_title="Delta",
                template="plotly_dark",
                hovermode="x unified",
                height=450
            )
            fig_delta.show()

            # --------------------------------------------------------
            # Graphique PnL : Yahoo vs Maison
            # --------------------------------------------------------
            fig_pnl = go.Figure()
            fig_pnl.add_trace(go.Scatter(
                x=results_df["step"],
                y=results_df["hedged_pnl_yahoo"],
                mode="lines+markers",
                name="PnL hedgé Yahoo",
                hovertemplate="Step: %{x}<br>PnL hedgé Yahoo: %{y:.6f}<extra></extra>"
            ))
            fig_pnl.add_trace(go.Scatter(
                x=results_df["step"],
                y=results_df["hedged_pnl_home"],
                mode="lines+markers",
                name="PnL hedgé Maison",
                hovertemplate="Step: %{x}<br>PnL hedgé Maison: %{y:.6f}<extra></extra>"
            ))
            fig_pnl.add_trace(go.Scatter(
                x=results_df["step"],
                y=results_df["unhedged_pnl_yahoo"],
                mode="lines",
                name="PnL non hedgé Yahoo",
                line=dict(dash="dash"),
                hovertemplate="Step: %{x}<br>PnL non hedgé Yahoo: %{y:.6f}<extra></extra>"
            ))
            fig_pnl.add_trace(go.Scatter(
                x=results_df["step"],
                y=results_df["unhedged_pnl_home"],
                mode="lines",
                name="PnL non hedgé Maison",
                line=dict(dash="dot"),
                hovertemplate="Step: %{x}<br>PnL non hedgé Maison: %{y:.6f}<extra></extra>"
            ))
            fig_pnl.update_layout(
                title="Comparaison PnL hedgé / non hedgé — Yahoo vs Maison",
                xaxis_title="Pas de temps",
                yaxis_title="PnL",
                template="plotly_dark",
                hovermode="x unified",
                height=500
            )
            fig_pnl.show()

            # --------------------------------------------------------
            # Tableau
            # --------------------------------------------------------
            display(HTML("<h4 style='margin: 16px 0 8px 0;'>Tableau de simulation du hedge : Yahoo vs Maison</h4>"))
            display(HTML(render_dataframe_html(results_df.round(6))))

        except Exception as e:
            display(HTML(f"<b style='color:#d62828;'>Erreur delta hedging :</b> {e}"))


# ------------------------------------------------------------
# UI section 4
# ------------------------------------------------------------
hedge_ui = widgets.VBox([
    widgets.HTML("<h3 style='margin:16px 0 6px 0;'>4. Delta hedging dynamique connecté au panneau marché</h3>"),
    widgets.HTML("<p style='margin:0 0 10px 0; color:#555;'>Cette section reprend automatiquement le ticker, le side, l’échéance et le strike sélectionnés au-dessus. La comparaison Yahoo vs Maison est faite directement sur la même trajectoire simulée.</p>"),
    widgets.HBox([hedge_steps_widget, hedge_qty_widget, hedge_mu_widget, hedge_seed_widget, hedge_run_button])
])

display(hedge_ui, hedge_output)

hedge_run_button.on_click(run_delta_hedge_simulation)

# Premier affichage
run_delta_hedge_simulation()
```
