## Fondements mathématiques

Le cadre de Black-Scholes repose sur l’hypothèse que le prix du sous-jacent $S_t$ suit un mouvement brownien géométrique :

$$
dS_t = \mu S_t\,dt + \sigma S_t\,dW_t
$$

où :

- $\mu$ désigne la dérive,
- $\sigma$ la volatilité,
- $W_t$ un mouvement brownien.

En appliquant le lemme d’Itô à la valeur d’une option $V(S,t)$, puis en construisant un portefeuille auto-financé sans risque, on obtient l’équation aux dérivées partielles de Black-Scholes :

$$
\frac{\partial V}{\partial t}
+ \frac{1}{2}\sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}
+ rS \frac{\partial V}{\partial S}
- rV = 0
$$

Sous des conditions terminales appropriées, cette équation conduit aux formules fermées des options européennes.

Le prix d’un call européen s’écrit :

$$
C(S_0,K,T,r,\sigma) = S_0 \Phi(d_1) - K e^{-rT}\Phi(d_2)
$$

avec :

$$
d_1 = \frac{\ln\left(\frac{S_0}{K}\right) + \left(r + \frac{\sigma^2}{2}\right)T}{\sigma\sqrt{T}}
$$

$$
d_2 = d_1 - \sigma\sqrt{T}
$$

Le prix d’un put européen est donné par :

$$
P(S_0,K,T,r,\sigma) = K e^{-rT}\Phi(-d_2) - S_0 \Phi(-d_1)
$$

# 1. Fondements mathématiques

## 1.1 Pourquoi un cadre mathématique pour les options ?

Le prix d’une option dépend de plusieurs variables qui évoluent dans le temps :

- le prix du sous-jacent,
- la volatilité,
- le temps restant jusqu’à l’échéance,
- le taux d’intérêt,
- et parfois les dividendes.

Pour construire un modèle cohérent, il faut décrire mathématiquement l’évolution du sous-jacent, puis déterminer comment cette évolution affecte la valeur de l’option.

L’objectif de cette section est de poser les bases théoriques du modèle de Black-Scholes, qui servira ensuite pour :

- le pricing,
- le calcul des Greeks,
- la volatilité implicite,
- le hedging dynamique,
- et le stress testing.

---

## 1.2 Dynamique du sous-jacent

Dans le modèle de Black-Scholes, on suppose que le prix du sous-jacent $S_t$ suit un **mouvement brownien géométrique** :

$$
dS_t = \mu S_t\,dt + \sigma S_t\,dW_t
$$

où :

- $\mu$ est la dérive du sous-jacent,
- $\sigma$ est la volatilité,
- $W_t$ est un mouvement brownien standard.

Cette équation signifie que la variation instantanée du sous-jacent possède :

- une partie **déterministe** : $\mu S_t\,dt$,
- une partie **aléatoire** : $\sigma S_t\,dW_t$.

Le caractère multiplicatif du modèle permet de conserver un prix strictement positif, ce qui est cohérent pour un actif financier.

---

## 1.3 Valeur d’une option

On note $V(S,t)$ la valeur d’une option à la date $t$, lorsque le sous-jacent vaut $S$.

Comme $V$ dépend à la fois :

- du temps,
- et du niveau du sous-jacent,

on peut appliquer le **lemme d’Itô** à la fonction $V(S_t,t)$.

On obtient alors :

$$
dV = \frac{\partial V}{\partial t}dt
+ \frac{\partial V}{\partial S}dS
+ \frac{1}{2}\frac{\partial^2 V}{\partial S^2}(dS)^2
$$

En utilisant la dynamique de $S_t$ et la relation $(dW_t)^2 = dt$, cela devient :

$$
dV =
\left(
\frac{\partial V}{\partial t}
+ \mu S \frac{\partial V}{\partial S}
+ \frac{1}{2}\sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}
\right) dt
+ \sigma S \frac{\partial V}{\partial S} dW_t
$$

Cette expression est essentielle : elle montre que la valeur de l’option évolue elle aussi de façon stochastique.

---

## 1.4 Idée du portefeuille sans risque

L’intuition fondamentale de Black-Scholes consiste à construire un portefeuille combinant :

- une option,
- et une certaine quantité du sous-jacent,

de manière à éliminer le terme aléatoire en $dW_t$.

On considère le portefeuille :

$$
\Pi = V - \Delta S
$$

où $\Delta$ représente la quantité de sous-jacent détenue en couverture.

En choisissant :

$$
\Delta = \frac{\partial V}{\partial S}
$$

le terme stochastique disparaît.  
Le portefeuille devient alors localement sans risque.

Dans un marché sans opportunité d’arbitrage, un portefeuille sans risque doit rapporter le taux sans risque $r$.

Cette idée mène directement à l’équation de Black-Scholes.

---

## 1.5 Équation de Black-Scholes

En imposant la condition d’absence d’arbitrage, on obtient l’équation aux dérivées partielles suivante :

$$
\frac{\partial V}{\partial t}
+ \frac{1}{2}\sigma^2 S^2 \frac{\partial^2 V}{\partial S^2}
+ rS \frac{\partial V}{\partial S}
- rV = 0
$$

Cette équation doit être satisfaite par toute option européenne dans le cadre du modèle de Black-Scholes.

Elle relie :

- la variation de la valeur dans le temps,
- la sensibilité au sous-jacent,
- la convexité,
- et le coût de financement au taux sans risque.

---

## 1.6 Conditions terminales

Pour résoudre l’équation de Black-Scholes, il faut spécifier le payoff à maturité.

Pour un **call européen** de strike $K$ et de maturité $T$ :

$$
V(S,T) = \max(S_T - K, 0)
$$

Pour un **put européen** :

$$
V(S,T) = \max(K - S_T, 0)
$$

À partir de ces conditions terminales, on obtient les formules fermées de Black-Scholes.

---

## 1.7 Formules de Black-Scholes

Le prix d’un call européen est donné par :

$$
C(S_0,K,T,r,\sigma) = S_0 \Phi(d_1) - K e^{-rT}\Phi(d_2)
$$

avec :

$$
d_1 = \frac{\ln\left(\frac{S_0}{K}\right) + \left(r + \frac{\sigma^2}{2}\right)T}{\sigma\sqrt{T}}
$$

$$
d_2 = d_1 - \sigma\sqrt{T}
$$

Le prix d’un put européen est :

$$
P(S_0,K,T,r,\sigma) = K e^{-rT}\Phi(-d_2) - S_0 \Phi(-d_1)
$$

où $\Phi$ désigne la fonction de répartition de la loi normale centrée réduite.

---

## 1.8 Intuition financière

Le modèle de Black-Scholes repose sur une idée très forte :

> il est possible de répliquer localement le risque d’une option à l’aide du sous-jacent.

Autrement dit, le prix d’une option n’est pas arbitraire : il découle d’un raisonnement de couverture et d’absence d’arbitrage.

C’est cette logique qui rend ensuite les **Greeks** si importants :

- le **Delta** vient naturellement de la couverture,
- le **Gamma** mesure la variation de cette couverture,
- le **Vega** capte l’effet de la volatilité,
- le **Theta** l’effet du temps,
- le **Rho** l’effet des taux.

Nous allons maintenant passer de ce cadre théorique à l’implémentation numérique du modèle.

# 2. Moteur de pricing Black-Scholes

Dans cette section, nous allons implémenter le moteur de pricing Black-Scholes pour les options européennes.

L’objectif est de construire des fonctions propres, réutilisables et robustes pour :

- calculer les quantités intermédiaires $d_1$ et $d_2$,
- valoriser un call européen,
- valoriser un put européen,
- préparer la suite du notebook, notamment le calcul des Greeks et la volatilité implicite.

Nous mettrons également en place une configuration graphique plus avancée afin d’obtenir des visualisations claires et professionnelles.

```python
# ============================================================
# 2.0 Imports et configuration de base
# ============================================================

# -------------------------
# Calcul numérique et data
# -------------------------
import numpy as np
import pandas as pd

# -------------------------
# Statistiques / distributions
# -------------------------
from scipy.stats import norm
from scipy.optimize import brentq

# -------------------------
# Widgets et affichage notebook
# -------------------------
import ipywidgets as widgets
from IPython.display import display, HTML, Markdown

# -------------------------
# Visualisation Plotly
# -------------------------
import plotly.graph_objects as go
import plotly.express as px
import plotly.io as pio
from plotly.subplots import make_subplots

# -------------------------
# Réglages d'affichage pandas
# -------------------------
pd.set_option("display.max_columns", 50)
pd.set_option("display.width", 140)
pd.set_option("display.float_format", lambda x: f"{x:,.6f}")

# -------------------------
# Template graphique Plotly
# -------------------------
pio.templates.default = "plotly_dark"

print("Imports et configuration chargés avec succès.")
```

```python
# ============================================================
# 2.1 Fonctions de pricing Black-Scholes
# ============================================================

def _to_numpy(x):
    """
    Convertit une valeur scalaire ou un array-like en tableau numpy float.
    """
    return np.asarray(x, dtype=float)


def compute_d1(S, K, T, r, sigma):
    """
    Calcule d1 dans le modèle de Black-Scholes.
    """
    S = _to_numpy(S)
    K = _to_numpy(K)
    T = _to_numpy(T)
    r = _to_numpy(r)
    sigma = _to_numpy(sigma)

    if np.any(S <= 0):
        raise ValueError("Le spot S doit être strictement positif.")
    if np.any(K <= 0):
        raise ValueError("Le strike K doit être strictement positif.")
    if np.any(T <= 0):
        raise ValueError("La maturité T doit être strictement positive.")
    if np.any(sigma <= 0):
        raise ValueError("La volatilité sigma doit être strictement positive.")

    return (np.log(S / K) + (r + 0.5 * sigma**2) * T) / (sigma * np.sqrt(T))


def compute_d2(S, K, T, r, sigma):
    """
    Calcule d2 dans le modèle de Black-Scholes.
    """
    return compute_d1(S, K, T, r, sigma) - sigma * np.sqrt(T)


def bs_call_price(S, K, T, r, sigma):
    """
    Prix d'un call européen dans le modèle de Black-Scholes.
    """
    d1 = compute_d1(S, K, T, r, sigma)
    d2 = compute_d2(S, K, T, r, sigma)
    return S * norm.cdf(d1) - K * np.exp(-r * T) * norm.cdf(d2)


def bs_put_price(S, K, T, r, sigma):
    """
    Prix d'un put européen dans le modèle de Black-Scholes.
    """
    d1 = compute_d1(S, K, T, r, sigma)
    d2 = compute_d2(S, K, T, r, sigma)
    return K * np.exp(-r * T) * norm.cdf(-d2) - S * norm.cdf(-d1)
```

```python
# ============================================================
# 2.2 Fonctions des Greeks Black-Scholes
# ============================================================

def call_delta(S, K, T, r, sigma):
    """
    Delta d'un call européen.
    """
    d1 = compute_d1(S, K, T, r, sigma)
    return norm.cdf(d1)


def put_delta(S, K, T, r, sigma):
    """
    Delta d'un put européen.
    """
    d1 = compute_d1(S, K, T, r, sigma)
    return norm.cdf(d1) - 1


def gamma_bs(S, K, T, r, sigma):
    """
    Gamma d'un call ou d'un put européen.
    Identique pour call et put dans Black-Scholes.
    """
    d1 = compute_d1(S, K, T, r, sigma)
    S = _to_numpy(S)
    sigma = _to_numpy(sigma)
    T = _to_numpy(T)

    return norm.pdf(d1) / (S * sigma * np.sqrt(T))


def vega_bs(S, K, T, r, sigma):
    """
    Vega d'un call ou d'un put européen.
    """
    d1 = compute_d1(S, K, T, r, sigma)
    S = _to_numpy(S)
    T = _to_numpy(T)

    return S * norm.pdf(d1) * np.sqrt(T)


def call_theta(S, K, T, r, sigma):
    """
    Theta d'un call européen.
    """
    d1 = compute_d1(S, K, T, r, sigma)
    d2 = compute_d2(S, K, T, r, sigma)

    S = _to_numpy(S)
    K = _to_numpy(K)
    T = _to_numpy(T)
    sigma = _to_numpy(sigma)

    term1 = -S * norm.pdf(d1) * sigma / (2 * np.sqrt(T))
    term2 = -r * K * np.exp(-r * T) * norm.cdf(d2)

    return term1 + term2


def put_theta(S, K, T, r, sigma):
    """
    Theta d'un put européen.
    """
    d1 = compute_d1(S, K, T, r, sigma)
    d2 = compute_d2(S, K, T, r, sigma)

    S = _to_numpy(S)
    K = _to_numpy(K)
    T = _to_numpy(T)
    sigma = _to_numpy(sigma)

    term1 = -S * norm.pdf(d1) * sigma / (2 * np.sqrt(T))
    term2 = r * K * np.exp(-r * T) * norm.cdf(-d2)

    return term1 + term2


def call_rho(S, K, T, r, sigma):
    """
    Rho d'un call européen.
    """
    d2 = compute_d2(S, K, T, r, sigma)
    K = _to_numpy(K)
    T = _to_numpy(T)

    return K * T * np.exp(-r * T) * norm.cdf(d2)


def put_rho(S, K, T, r, sigma):
    """
    Rho d'un put européen.
    """
    d2 = compute_d2(S, K, T, r, sigma)
    K = _to_numpy(K)
    T = _to_numpy(T)

    return -K * T * np.exp(-r * T) * norm.cdf(-d2)
```
